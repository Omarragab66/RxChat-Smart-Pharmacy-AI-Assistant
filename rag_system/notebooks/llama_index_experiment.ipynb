```python
df = pd.read_csv("core_data.csv", engine="python")
df.head()
```

```python
drop_col = [
"drug_id",
"route",
"rxnorm_id",
"pubchem_cid",
"data_quality_score",
"llm_fields_count",
"is_egyptian",
"sources",
"last_updated",
"egyptian_price",
"top_adverse_events"
]
df.drop(columns=drop_col,inplace=True)
```

```python
# اعمل rag_text تلقائيًا من الأعمدة الموجودة
df['rag_text'] = df['rag_text'].fillna(
    df['drug_name'] + " - " + df['generic_name'].fillna("")
)

print(f"بعد الملء: {df['rag_text'].notna().sum()} صف")
```

```python
import pandas as pd
from llama_index.core import Document

documents = []

for _, row in df.iterrows():
    if pd.notna(row["rag_text"]):
        # ادمج اسم الدواء في النص عشان الـ embedding يشوفه
        full_text = f"اسم الدواء: {row['drug_name']}\n"
        full_text += f"الاسم الجنيسي: {row['generic_name']}\n"
        full_text += f"المعلومات: {row['rag_text']}"
        
        documents.append(
            Document(
                text=full_text,
                metadata={
                    "drug_name": row["drug_name"],
                    "generic_name": row["generic_name"]
                }
            )
        )

print(f"عدد الـ documents: {len(documents)}")
print("\nمثال على أول document:")
print(documents[0].text[:300])
```

```python
from llama_index.core.node_parser import SentenceSplitter

node_parser=SentenceSplitter(chunk_size=1024,chunk_overlap=100)
```

```python
nodes=node_parser.get_nodes_from_documents(documents=documents,show_progress=False)
```

```python
from llama_index.core import VectorStoreIndex ,StorageContext
from llama_index.vector_stores.chroma import ChromaVectorStore
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

```python
# خيار 1 (موصى به)
embed_model = HuggingFaceEmbedding(
    model_name="intfloat/multilingual-e5-large"
)

# خيار 2
# model_name="intfloat/multilingual-e5-large"   # كبير شوية لكنه قوي
```

```python
import chromadb


db = chromadb.PersistentClient(path="chroma_db_test")
print("Collections موجودة:", db.list_collections())

# امسح وتأكد
db.delete_collection("chroma_db_test")
print("Collections بعد المسح:", db.list_collections())
chroma_collection = db.get_or_create_collection("chroma_db_test")

vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)

# أعد بناء الـ index من أول
index = VectorStoreIndex(
    nodes=nodes,
    embed_model=embed_model,
    storage_context=storage_context
)
```

```python
retriever=index.as_retriever(similarity_top_k=3)
retrieved_nodes=retriever.retrieve("ما هوا دواء بانادول نايت")
```

```python
from llama_index.llms.ollama import Ollama
from llama_index.core.response_synthesizers import get_response_synthesizer

llm = Ollama(model="tinyllama", request_timeout=120.0)

retriever = index.as_retriever()
retrieved_nodes = retriever.retrieve("ما هوا دواء بانادول اطفال")

response_synthesizer = get_response_synthesizer(llm=llm)

response = response_synthesizer.synthesize(
    query="ما هوا دواء بانادول اطفال",
    nodes=retrieved_nodes
)

print(str(response))
```

```python
from llama_index.core import PromptTemplate
qa_prompt = PromptTemplate(
    """أنت مساعد صيدلاني ذكي واسمك "دكتور أدهم". شخصيتك ودية ومحبوبة وبتتكلم بأسلوب بسيط ومفهوم.

    قواعدك:
    - استخدم فقط المعلومات المتاحة للإجابة
    - لو مش لاقي المعلومة قول "معنديش معلومة كافية عن ده دلوقتي"
    - لا تخترع معلومات
    - اتكلم بالعربي دايماً
    - ردودك تكون مفصلة ومنظمة (مش سطر واحد)

    أسلوب ردك يكون:
    1. ابدأ بتحية أو جملة ترحيب قصيرة
    2. اشرح الدواء بالتفصيل (الاستخدام، المادة الفعالة، الجرعة، التحذيرات)
    3. استخدم نقاط أو تنسيق واضح
    4. في الآخر اسأل سؤال يشجع المحادثة، مثل:
       "هل عندك استفسار عن الجرعة؟" أو "هل الدواء ده لحد كبير ولا طفل؟"

    المعلومات المتاحة:
    {context_str}

    السؤال: {query_str}
    
    الإجابة:"""
)
query_engine = index.as_query_engine(
    llm=Ollama(model="mistral", request_timeout=300.0),
    text_qa_template=qa_prompt,
    similarity_top_k=4
)

response = query_engine.query("اشرحلي دواء بانادول اطفال بالتفصيل")
print(str(response))
```
