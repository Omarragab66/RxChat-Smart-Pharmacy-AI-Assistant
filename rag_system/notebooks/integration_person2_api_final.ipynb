## 🧪 Cell 3 — Person 2: Connect Safety API (Flask)

هذه الخلية تجعل النوتبوك يستخدم نفس الـ API التي تعمل على `http://127.0.0.1:5000/predict` بدل تحميل موديل مختلف داخل النوتبوك.


````python
# ============================================================
# Person 2 — Safety Classifier API Adapter
# ============================================================
# مهم: شغّل Flask API بتاعة Person 2 الأول.
# من الصورة عندك هي شغالة هنا:
# http://127.0.0.1:5000/predict?text=...
#
# السبب القديم للمشكلة:
# النوتبوك كان بيحمّل موديل محليًا، بينما الـ API شغالة بموديل/بايبلاين تانية.
# لذلك النتيجة كانت مختلفة. هنا هنخلي النوتبوك ينادي نفس الـ API.

import re
import ast
import html
import requests

PERSON2_API_URL = "http://127.0.0.1:5000/predict"


def _clean_plain_text(x: str) -> str:
    """Convert HTML response to readable plain text."""
    x = html.unescape(x or "")
    x = re.sub(r"<script.*?</script>", " ", x, flags=re.DOTALL | re.IGNORECASE)
    x = re.sub(r"<style.*?</style>", " ", x, flags=re.DOTALL | re.IGNORECASE)
    x = re.sub(r"<[^>]+>", " ", x)
    x = re.sub(r"\s+", " ", x).strip()
    return x


def _normalize_label(label) -> str:
    """Normalize label from API to: Low / Medium / High."""
    label = str(label or "").strip().lower()

    mapping = {
        "low": "Low",
        "منخفض": "Low",
        "آمن": "Low",
        "امن": "Low",

        "medium": "Medium",
        "med": "Medium",
        "متوسط": "Medium",
        "احتياط": "Medium",

        "high": "High",
        "خطر": "High",
        "عالي": "High",
    }

    return mapping.get(label, str(label).capitalize() if label else "Medium")


def _normalize_probabilities(probs, label="Medium", confidence=None) -> dict:
    """Normalize probability keys from API/HTML to lowercase keys used by pipeline."""
    probs = probs or {}

    def pick(*keys):
        for k in keys:
            if k in probs:
                try:
                    return float(probs[k])
                except Exception:
                    return 0.0
        return 0.0

    normalized = {
        "low": pick("Low", "low", "LOW", 0, "0"),
        "medium": pick("Medium", "medium", "MEDIUM", 1, "1"),
        "high": pick("High", "high", "HIGH", 2, "2"),
    }

    # لو الـ API رجعت confidence فقط بدون probabilities
    if sum(normalized.values()) == 0 and confidence is not None:
        try:
            conf = float(confidence)
        except Exception:
            conf = 0.0

        label = _normalize_label(label)
        remain = max(0.0, 1.0 - conf)
        normalized = {"low": remain / 2, "medium": remain / 2, "high": remain / 2}
        normalized[label.lower()] = conf

    return {k: round(float(v), 4) for k, v in normalized.items()}


def _build_safety_result(label, confidence=None, probabilities=None) -> dict:
    """Build output with same shape used by the old notebook pipeline."""
    risk_level = _normalize_label(label)
    probabilities = _normalize_probabilities(probabilities, risk_level, confidence)

    high_prob = probabilities.get("high", 0.0)

    if risk_level == "High":
        flags = ["DANGEROUS", "CONSULT_DOCTOR"]
        explanation = "⚠️ خطر طبي عالي — يرجى استشارة الطبيب فوراً"
    elif risk_level == "Medium":
        flags = ["CAUTION_REQUIRED"]
        explanation = "⚡ احتياط مطلوب — راجع الصيدلاني"
    else:
        flags = []
        explanation = "✅ آمن — لا توجد مخاطر واضحة"

    return {
        "risk_level": risk_level,
        "safety_score": round(float(high_prob), 4),
        "risk_flags": flags,
        "explanation": explanation,
        "probabilities": probabilities,
        "source": "person2_flask_api",
    }


def _parse_api_json(data: dict) -> dict:
    """
    Parse Person 2 Flask API response.

    Expected API shape:
    {
        "input_text": "...",
        "prediction": {
            "risk_level": "Low",
            "confidence": 0.98,
            "probabilities": {
                "Low": 0.98,
                "Medium": 0.01,
                "High": 0.01
            }
        }
    }
    """

    # أهم إصلاح:
    # لو النتيجة جوه prediction كـ dict، ادخل جواها
    if isinstance(data.get("prediction"), dict):
        data = data["prediction"]

    label = (
        data.get("risk_level")
        or data.get("label")
        or data.get("classification")
        or data.get("class")
    )

    confidence = (
        data.get("confidence")
        or data.get("score")
        or data.get("probability")
    )

    probabilities = (
        data.get("probabilities")
        or data.get("probs")
        or data.get("scores")
        or {}
    )

    return _build_safety_result(label, confidence, probabilities)

def _parse_api_html(html_text: str) -> dict:
    """Parse the current browser HTML page returned by /predict?text=..."""
    plain = _clean_plain_text(html_text)

    # Label after Arabic/English title, fallback to first Low/Medium/High found
    label_match = re.search(
        r"(?:التصنيف|classification|prediction|label)\s*:?\s*(Low|Medium|High)",
        plain,
        flags=re.IGNORECASE,
    )
    if not label_match:
        label_match = re.search(r"\b(Low|Medium|High)\b", plain, flags=re.IGNORECASE)

    label = label_match.group(1) if label_match else "Medium"

    conf_match = re.search(r"Confidence\s*:?\s*([0-9]*\.?[0-9]+)", plain, flags=re.IGNORECASE)
    confidence = float(conf_match.group(1)) if conf_match else None

    probabilities = {}
    dict_match = re.search(r"\{[^{}]*(?:'Low'|\"Low\"|low)[^{}]*\}", plain, flags=re.IGNORECASE)
    if dict_match:
        try:
            probabilities = ast.literal_eval(dict_match.group(0))
        except Exception:
            probabilities = {}

    return _build_safety_result(label, confidence, probabilities)


def classify_safety(text: str) -> dict:
    """
    Person 2: classify query risk using the running Flask API.
    Returns same keys expected by rxchat_pipeline.
    """
    errors = []

    # First try POST JSON, in case the API supports it
    try:
        response = requests.post(
            PERSON2_API_URL,
            json={"text": text, "query": text},
            timeout=20,
            headers={"Accept": "application/json"},
        )

        if response.status_code == 200:
            ctype = response.headers.get("content-type", "").lower()
            if "application/json" in ctype:
                return _parse_api_json(response.json())
            return _parse_api_html(response.text)

        errors.append(f"POST {response.status_code}: {response.text[:150]}")
    except Exception as e:
        errors.append(f"POST error: {e}")

    # Fallback to the browser-style GET endpoint shown in your screenshot
    try:
        response = requests.get(
            PERSON2_API_URL,
            params={"text": text},
            timeout=20,
            headers={"Accept": "application/json"},
        )

        if response.status_code == 200:
            ctype = response.headers.get("content-type", "").lower()
            if "application/json" in ctype:
                return _parse_api_json(response.json())
            return _parse_api_html(response.text)

        errors.append(f"GET {response.status_code}: {response.text[:150]}")
    except Exception as e:
        errors.append(f"GET error: {e}")

    raise RuntimeError(
        "Person 2 API connection failed. تأكد إن Flask API شغالة على "
        f"{PERSON2_API_URL}. Details: " + " | ".join(errors)
    )


# Quick test — لازم يرجع نفس نتيجة الصفحة اللي في المتصفح
try:
    test = classify_safety("ايه الاعراض الجانبية لكوتكورد؟")
    print(f"✅ Person 2 API connected: {test['risk_level']} | {test['probabilities']}")
except Exception as e:
    print("❌ Person 2 API test failed")
    print(e)
````

## 💊 Cell 4 — Person 3: Load Dosage Calculator

````python
import json
import random
import re
import joblib
from groq import Groq
import requests
print("⏳ Loading Dosage Calculator (Groq reasoning + Rule-based)...")

# Load dosage data for few-shot
with open(PATHS["dosage_json"], "r", encoding="utf-8") as f:
    dosage_data = json.load(f)

# Load or build safety classifier
def build_features_simple(age, weight, gender, conditions, kidney, liver):
    kidney_map = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    liver_map  = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    return [
        age, weight,
        1 if gender == "ذكر" else 0,
        len(conditions),
        kidney_map.get(kidney, 0),
        liver_map.get(liver, 0),
        1 if any("سكري" in c for c in conditions) else 0,
        1 if any("قلب" in c or "ضغط" in c for c in conditions) else 0,
        1 if any("كبد" in c for c in conditions) else 0,
        1 if any("كلى" in c or "كلوي" in c for c in conditions) else 0,
    ]

def rule_based_risk(age, conditions, kidney, liver):
    """Rule-based risk — Person 3 safety component"""
    kidney_map = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    liver_map  = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    k = kidney_map.get(kidney, 0)
    l = liver_map.get(liver, 0)
    n = len(conditions)
    if k >= 3 or l >= 3 or n >= 3:
        return "HIGH"
    elif k >= 1 or l >= 1 or n >= 1 or age >= 65 or age <= 12:
        return "MEDIUM"
    return "LOW"


def clean_llm_text(text: str) -> str:
    """Remove <think> blocks before showing LLM output to the user."""
    text = re.sub(r"<think>.*?</think>", "", text or "", flags=re.DOTALL | re.IGNORECASE)
    return text.strip()


def extract_line_value(lines, keys):
    for line in lines:
        if any(k in line for k in keys):
            return line.split(":", 1)[-1].strip() if ":" in line else line.strip()
    return ""

def calculate_dosage(drug_name: str, patient: dict) -> dict:
    """
    Person 4 Orchestrator:
    يرسل drug_name + patient_profile إلى Person 3 Dosage API
    ويستقبل نتيجة الجرعة.
    """

    url = "http://127.0.0.1:8003/dosage"

    payload = {
        "drug_name": drug_name,
        "patient_profile": {
            "age": patient.get("age"),
            "weight_kg": patient.get("weight_kg"),
            "gender": patient.get("gender", "ذكر"),
            "chronic_conditions": patient.get("chronic_conditions", []),
            "kidney_function": patient.get("kidney_function", "normal"),
            "liver_function": patient.get("liver_function", "normal")
        }
    }

    try:
        response = requests.post(url, json=payload, timeout=30)

        if response.status_code != 200:
            return {
                "drug_name": drug_name,
                "risk_category": "UNKNOWN",
                "recommended_dose": "",
                "reasoning": f"Dosage API returned error {response.status_code}: {response.text}",
                "chain_of_thought": [],
                "dose_adjustment_reason": "",
                "safety_flag": True,
                "doctor_referral": True
            }

        result = response.json()

        return {
            "drug_name": result.get("drug_name", drug_name),
            "risk_category": result.get("risk_category", "UNKNOWN"),
            "recommended_dose": result.get("recommended_dose", ""),
            "reasoning": result.get("reasoning", ""),
            "chain_of_thought": result.get("chain_of_thought", []),
            "dose_adjustment_reason": result.get("dose_adjustment_reason", ""),
            "safety_flag": result.get("safety_flag", False),
            "doctor_referral": result.get("doctor_referral", False)
        }

    except Exception as e:
        return {
            "drug_name": drug_name,
            "risk_category": "UNKNOWN",
            "recommended_dose": "",
            "reasoning": f"Dosage API connection failed: {str(e)}",
            "chain_of_thought": [],
            "dose_adjustment_reason": "",
            "safety_flag": True,
            "doctor_referral": True
        }

print(f"✅ Dosage engine ready — {len(dosage_data)} examples loaded")
````

## 📚 Cell 5 — Person 4: Load RAG Pipeline

````python
import re
import os
import numpy as np
import pandas as pd
import requests
from rapidfuzz import process, fuzz
from llama_index.core import Document, VectorStoreIndex, Settings, StorageContext
from llama_index.core.node_parser import SentenceSplitter
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core.vector_stores import MetadataFilters, MetadataFilter
from llama_index.vector_stores.chroma import ChromaVectorStore
import chromadb

print("⏳ Loading RAG Pipeline...")

# Load CSV
df = pd.read_csv(PATHS["core_csv"])
df.columns = [str(c).strip() for c in df.columns]

needed_cols = [
    'drug_id', 'drug_name', 'generic_name', 'drug_class', 'form', 'route',
    'indications', 'dosage', 'mechanism', 'warnings', 'contraindications',
    'drug_interactions', 'adverse_reactions', 'top_adverse_events',
    'pregnancy_info', 'pediatric_info', 'geriatric_info', 'overdose_info',
    'storage_info', 'sources', 'is_egyptian', 'last_updated'
]
for col in needed_cols:
    if col not in df.columns:
        df[col] = ''

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('').astype(str).str.strip()

print(f"✅ CSV loaded: {len(df)} drugs")


# ── Build structured section documents ──
def clean_text(x):
    x = str(x).strip()
    return re.sub(r'\s+', ' ', x) if x.lower() not in {'nan','none',''} else ''

def section_text(title_ar, title_en, content):
    c = clean_text(content)
    return f"{title_ar} / {title_en}:{c}" if c else ''

SECTIONS_MAP = [
    ('indications',        'الاستخدامات',           'Indications'),
    ('dosage',             'الجرعة',                'Dosage'),
    ('mechanism',          'آلية العمل',             'Mechanism'),
    ('warnings',           'التحذيرات',              'Warnings'),
    ('contraindications',  'موانع الاستخدام',        'Contraindications'),
    ('drug_interactions',  'التفاعلات الدوائية',     'Drug interactions'),
    ('adverse_reactions',  'الأعراض الجانبية',       'Adverse reactions'),
    ('top_adverse_events', 'أشهر البلاغات الجانبية', 'Top FAERS events'),
    ('pregnancy_info',     'الحمل والرضاعة',         'Pregnancy'),
    ('pediatric_info',     'الأطفال',                'Pediatric'),
    ('geriatric_info',     'كبار السن',              'Geriatric'),
    ('overdose_info',      'الجرعة الزائدة',         'Overdose'),
    ('storage_info',       'التخزين',                'Storage'),
]

documents = []
for _, row in df.iterrows():
    dname   = clean_text(row.get('drug_name', ''))
    generic = clean_text(row.get('generic_name', ''))

    header = f"""اسم الدواء: {dname} | Generic: {generic}
الفئة: {clean_text(row.get('drug_class',''))} | الشكل: {clean_text(row.get('form',''))}
Route: {clean_text(row.get('route',''))} | مصري: {'نعم' if row.get('is_egyptian')=='1' else 'غير مؤكد'}"""

    meta = {'drug_name': dname, 'generic_name': generic,
            'drug_class': clean_text(row.get('drug_class','')), 'section': 'overview'}
    documents.append(Document(
        text=f"دواء: {dname} | Generic: {generic} | Section: overview{header}",
        metadata=meta
    ))

    for col, ar, en in SECTIONS_MAP:
        text = section_text(ar, en, row.get(col, ''))
        if text:
            m = {**meta, 'section': col}
            documents.append(Document(
                text=f"دواء: {dname} | Generic: {generic} | Section: {col}{text}",
                metadata=m
            ))

print(f"✅ Built {len(documents)} section documents")


# ── Person 1 adapter placeholder ──
def person1_nlp_adapter(query: str) -> dict:
    """
    مدخل جاهز لتركيب Person 1 لاحقاً.
    حالياً يعمل skip ويرجع query كما هي، بدون ما يكسر الـ pipeline.
    لما Person 1 يخلص، خليه يرجع:
    normalized_query, intent, drug_entities, corrected_drugs
    """
    if not USE_PERSON1_NLP:
        return {
            "enabled": False,
            "status": "skipped",
            "normalized_query": query,
            "intent": None,
            "drug_entities": [],
            "corrected_drugs": []
        }

    # TODO: later connect Person 1 API/model here
    # مثال:
    # resp = requests.post("http://127.0.0.1:8001/nlp", json={"query": query}).json()
    # return resp
    raise NotImplementedError("Person 1 NLP adapter is enabled but not connected yet.")


# ── Drug name matching fallback until Person 1 is ready ──
def normalize_ar(text):
    text = str(text).lower().strip()
    for old, new in [('[إأآا]','ا'),('ى','ي'),('ؤ','و'),('ئ','ي'),('ة','ه'),('[ًٌٍَُِّْـ]','')]:
        text = re.sub(old, new, text)
    return re.sub(r'\s+', ' ', re.sub(r'[^\w\s؀-ۿ]', ' ', text)).strip()

name_rows = []
for _, row in df.iterrows():
    for field in ['drug_name', 'generic_name']:
        val = clean_text(row.get(field, ''))
        if val:
            name_rows.append({
                'candidate': val,
                'candidate_norm': normalize_ar(val),
                'drug_name': clean_text(row.get('drug_name', '')),
                'generic_name': clean_text(row.get('generic_name', '')),
            })

names_df = pd.DataFrame(name_rows).drop_duplicates(['candidate_norm', 'drug_name'])
choices  = names_df['candidate_norm'].tolist()


def detect_drug(query: str, threshold: int = 78) -> dict:
    q = normalize_ar(query)
    for _, r in names_df.iterrows():
        c = r['candidate_norm']
        if c and (c in q or q in c):
            return {'matched': True, 'drug_name': r['drug_name'],
                    'generic_name': r['generic_name'], 'score': 100, 'method': 'exact'}
    match = process.extractOne(q, choices, scorer=fuzz.WRatio)
    if match:
        norm, score, idx = match
        r = names_df.iloc[idx]
        if score >= threshold:
            return {'matched': True, 'drug_name': r['drug_name'],
                    'generic_name': r['generic_name'], 'score': round(float(score),2), 'method': 'fuzzy'}
    return {'matched': False, 'drug_name': None, 'generic_name': None, 'score': 0, 'method': 'none'}


# ── Build persistent ChromaDB vector index ──
print("⏳ Building/loading persistent ChromaDB vector index...")
Settings.embed_model = HuggingFaceEmbedding(model_name=EMBED_MODEL)
Settings.llm         = None
Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=80)

chroma_client = chromadb.PersistentClient(path=PATHS["chroma_dir"])
chroma_collection = chroma_client.get_or_create_collection("rxchat_core_drugs")
vector_store = ChromaVectorStore(chroma_collection=chroma_collection)
storage_context = StorageContext.from_defaults(vector_store=vector_store)

if chroma_collection.count() == 0:
    print("  → Chroma collection is empty; indexing documents now...")
    rag_index = VectorStoreIndex.from_documents(
        documents,
        storage_context=storage_context,
        show_progress=True
    )
else:
    print(f"  → Existing Chroma collection found: {chroma_collection.count()} vectors")
    rag_index = VectorStoreIndex.from_vector_store(vector_store)

print("✅ ChromaDB RAG index ready!")


# ── Retrieval function ──
def retrieve_context(query: str, top_k: int = 5) -> tuple:
    drug_info = detect_drug(query)

    # If drug matched, use metadata filter inside ChromaDB.
    # If filtering fails for any version reason, fallback to normal retrieval.
    try:
        if drug_info['matched']:
            filters = MetadataFilters(filters=[
                MetadataFilter(key="drug_name", value=drug_info['drug_name'])
            ])
            retriever = rag_index.as_retriever(similarity_top_k=top_k, filters=filters)
        else:
            retriever = rag_index.as_retriever(similarity_top_k=top_k)
    except Exception:
        retriever = rag_index.as_retriever(similarity_top_k=top_k)

    nodes = retriever.retrieve(query)
    results = [{
        'score':        float(n.score) if n.score else 0,
        'drug_name':    n.node.metadata.get('drug_name'),
        'generic_name': n.node.metadata.get('generic_name'),
        'section':      n.node.metadata.get('section'),
        'text':         n.node.text,
    } for n in nodes]

    return drug_info, results


# ── Tavily Web Fallback ──
def rag_is_weak(rag_results: list, threshold: float = 0.35) -> bool:
    if not rag_results:
        return True
    best_score = max([r.get('score', 0) for r in rag_results] or [0])
    return best_score < threshold


def web_fallback_search(query: str, max_results: int = 3) -> dict:
    """
    Fallback search for unknown/weak RAG cases.
    لو TAVILY_API_KEY مش موجود، يرجع skipped بدون هلوسة.
    """
    if not TAVILY_API_KEY:
        return {
            "used": False,
            "reason": "TAVILY_API_KEY not set",
            "results": []
        }

    try:
        resp = requests.post(
            "https://api.tavily.com/search",
            json={
                "api_key": TAVILY_API_KEY,
                "query": query,
                "search_depth": "basic",
                "max_results": max_results,
                "include_answer": False,
            },
            timeout=15,
        )
        resp.raise_for_status()
        data = resp.json()
        return {
            "used": True,
            "reason": "weak_or_missing_rag",
            "results": data.get("results", [])[:max_results]
        }
    except Exception as e:
        return {
            "used": False,
            "reason": f"fallback_error: {e}",
            "results": []
        }


# Quick test
di, res = retrieve_context("ما هي جرعة بانادول اطفال؟")
print(f"🧪 Drug detected: {di['drug_name']} | Sections: {[r['section'] for r in res]}")
````

## 🖼️ Cell 6 — Image Understanding (Groq Vision)

````python
import base64
from pathlib import Path
from groq import Groq


def image_to_base64(image_path: str) -> str:
    """تحويل الصورة لـ base64"""
    with open(image_path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")


def identify_drug_from_image(image_path: str) -> dict:
    """
    Person 4 — Vision Module
    يستقبل صورة دواء → يرجع اسم الدواء
    """
    ext = Path(image_path).suffix.lower()
    mime_map = {'.jpg': 'image/jpeg', '.jpeg': 'image/jpeg',
                '.png': 'image/png', '.webp': 'image/webp'}
    mime = mime_map.get(ext, 'image/jpeg')

    img_b64 = image_to_base64(image_path)

    groq_client = Groq(api_key=GROQ_API_KEY)
    resp = groq_client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",  # Groq Vision model
        messages=[{
            "role": "user",
            "content": [
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:{mime};base64,{img_b64}"}
                },
                {
                    "type": "text",
                    "text": """أنت مساعد صيدلي.
انظر لصورة الدواء وأخبرني:
1. اسم الدواء التجاري
2. الاسم العلمي (generic name) إن ظهر
3. التركيز / الجرعة المكتوبة على العبوة إن ظهر

أجب بـ JSON فقط:
{"drug_name": "...", "generic_name": "...", "concentration": "...", "confidence": "high/medium/low"}"""
                }
            ]
        }],
        temperature=0.1,
        max_tokens=200,
    )

    raw = resp.choices[0].message.content.strip()

    # Parse JSON
    try:
        # Remove markdown code block if present
        raw_clean = re.sub(r'```json|```', '', raw).strip()
        result = json.loads(raw_clean)
    except:
        result = {'drug_name': raw, 'generic_name': '', 'concentration': '', 'confidence': 'low'}

    result['raw_response'] = raw
    return result


print("✅ Vision module ready")
print("\nUsage example:")
print('  result = identify_drug_from_image("path/to/drug_image.jpg")')
print('  print(result)')
````

## 🧠 Cell 7 — Main Pipeline: Everything Together

````python
SYSTEM_PROMPT = """
أنت RxChat، مساعد صيدلي ذكي يتكلم مع المستخدم بطريقة ودودة وواضحة بالعربية المصرية البسيطة.

شخصيتك:
- اتكلم كأنك فاهم حالة المستخدم، مش كأنك بتقرأ تقرير ثابت.
- ابدأ بتلخيص بسيط لحالة المريض لو بياناته موجودة.
- خلي ردك طبيعي ومباشر، ومش لازم تستخدم نفس العناوين كل مرة.
- اسأل 2 إلى 3 أسئلة متابعة ذكية لو الحالة محتاجة معلومات أكثر.
- استخدم أسلوب هادي ومطمئن، خصوصًا لو الحالة فيها خطورة.

قواعد مهمة:
- استخدم معلومات RAG و Safety و Dosage فقط.
- لا تخترع جرعات أو معلومات غير موجودة في السياق.
- لو نتيجة Safety = High أو doctor_referral = True، لا تعرض الجرعة كأنها تصريح بالاستخدام.
- في الحالات الخطرة، اذكر الجرعة إن كانت موجودة على أنها "نتيجة مبدئية من النظام"، لكن وضح أن الاستخدام يحتاج طبيب أو صيدلي.
- لا تقل للمستخدم "خد الدواء" في الحالات عالية الخطورة.
- لا تعطي جرعة إذا بيانات العمر أو الوزن أو وظائف الكلى والكبد ناقصة.
- لا تكرر الـ disclaimer.
- لا تستخدم قالب ثابت مثل: الاستخدام / الجرعة / التحذيرات / متى تذهب للطبيب في كل مرة.
- خلي الرد قصير، مفيد، وواضح.
""".strip()


def get_missing_dosage_info(drug_info: dict, patient_profile: dict) -> list:
    """Check missing required fields before calling Person 3 dosage model."""
    patient_profile = patient_profile or {}
    missing = []

    if not drug_info or not drug_info.get('matched'):
        missing.append("اسم الدواء")

    if patient_profile.get("age") in [None, "", 0]:
        missing.append("العمر")

    if patient_profile.get("weight_kg") in [None, "", 0]:
        missing.append("الوزن بالكيلو")

    if not patient_profile.get("kidney_function"):
        missing.append("وظائف الكلى")

    if not patient_profile.get("liver_function"):
        missing.append("وظائف الكبد")

    # chronic_conditions may be empty list; ده مقبول لو المريض لا يعاني من أمراض
    if "chronic_conditions" not in patient_profile:
        missing.append("هل توجد أمراض مزمنة؟")

    return missing


def build_missing_info_response(missing: list) -> str:
    questions_map = {
        "اسم الدواء": "اسم الدواء إيه؟",
        "العمر": "عمر المريض كام سنة؟",
        "الوزن بالكيلو": "وزن المريض كام كيلو؟",
        "وظائف الكلى": "هل وظائف الكلى طبيعية ولا فيه قصور؟",
        "وظائف الكبد": "هل وظائف الكبد طبيعية ولا فيه قصور؟",
        "هل توجد أمراض مزمنة؟": "هل عند المريض أمراض مزمنة؟ لو لا، اكتب: لا يوجد."
    }

    qs = [questions_map[x] for x in missing if x in questions_map]

    return (
        "تمام، عشان أحسب الجرعة بشكل آمن ومخمنش أي معلومة، محتاج أعرف شوية بيانات الأول:\n\n"
        + "\n".join([f"- {q}" for q in qs])
    )


def add_final_disclaimer(final_response: str, high_risk: bool = False) -> str:
    """Add warning/disclaimer with clean formatting."""
    final_response = final_response.strip()

    if high_risk:
        final_response += (
            "\n\n⚠️ **تنبيه مهم:** الحالة دي محتاجة مراجعة طبيب أو صيدلي قبل استخدام الدواء."
        )

    final_response += (
        "\n\n---\n"
        "*المعلومات تعليمية فقط وليست بديلاً عن الطبيب أو الصيدلي.*"
    )

    return final_response


def rxchat_pipeline(
    query: str,
    patient_profile: dict = None,
    image_path: str = None,
    verbose: bool = True
) -> dict:
    """
    🔗 Main RxChat Integration Pipeline
    =====================================
    Input:  query + optional patient_profile + optional image_path
    Output: final_response + all intermediate results
    """
    patient_profile = patient_profile or {}
    results = {}

    print("=" * 70)
    print(f"🔎 Query: {query}")
    if patient_profile:
        print(f"👤 Patient: {patient_profile}")
    print("=" * 70)

    # ─────────────────────────────────────────
    # STEP 0 (optional): Image → Drug Name
    # ─────────────────────────────────────────
    if image_path:
        print("📷 Step 0: Image Understanding...")
        vision_result = identify_drug_from_image(image_path)
        results['vision'] = vision_result

        drug_from_img = vision_result.get('drug_name', '')
        if drug_from_img and drug_from_img.lower() not in query.lower():
            query = f"{query} [{drug_from_img}]"

        if verbose:
            print(f"  → Drug identified: {vision_result.get('drug_name')}")
            print(f"  → Generic: {vision_result.get('generic_name')}")
            print(f"  → Confidence: {vision_result.get('confidence')}")

    # ─────────────────────────────────────────
    # STEP 1: Person 1 NLP Adapter — skipped now, ready later
    # ─────────────────────────────────────────
    print("🗣️ Step 1: Person 1 NLP Adapter...")
    p1_result = person1_nlp_adapter(query)
    results['person1_nlp'] = p1_result
    working_query = p1_result.get("normalized_query") or query

    if verbose:
        print(f"  → Person 1 status: {p1_result.get('status')}")

    # Temporary drug detection until Person 1 is connected
    print("🔍 Step 1b: Drug Detection Fallback...")
    detect_query = working_query

    p1_drugs = (
        p1_result.get("corrected_drugs")
        or p1_result.get("drug_entities")
        or []
    )

    if p1_drugs:
        detect_query = working_query + " " + " ".join(p1_drugs)

    drug_info = detect_drug(detect_query)
    results['drug_detection'] = drug_info

    if verbose:
        if drug_info['matched']:
            print(
                f"  → ✅ Detected: {drug_info['drug_name']} "
                f"(score={drug_info['score']}, method={drug_info['method']})"
            )
        else:
            print("  → ⚠️  No specific drug detected — global search/fallback may be used")

    # ─────────────────────────────────────────
    # Early Missing Info Check for dosage questions
    # ─────────────────────────────────────────
    dosage_keywords = ['جرعة', 'dose', 'كمية', 'كام حبة', 'مرات', 'مقدار']

    p1_intent = p1_result.get("intent")

    is_dosage_question = (
        p1_intent == "dosage"
        or any(kw in working_query.lower() for kw in dosage_keywords)
    )

    if is_dosage_question:
        missing = get_missing_dosage_info(drug_info, patient_profile)
        if missing:
            final_response = build_missing_info_response(missing)
            final_response = final_response

            results['status'] = 'need_more_info'
            results['missing_info'] = missing
            results['final_response'] = final_response

            print("✅ NEED MORE INFO RESPONSE:")
            print(final_response)
            return results

    # ─────────────────────────────────────────
    # STEP 2: RAG Retrieval (Person 4)
    # ─────────────────────────────────────────
    print("📚 Step 2: RAG Retrieval...")
    _, rag_results = retrieve_context(working_query)
    results['rag'] = rag_results

    if verbose:
        for i, r in enumerate(rag_results, 1):
            print(f"  [{i}] {r['drug_name']} / {r['section']} (score={r['score']:.3f})")

    # ─────────────────────────────────────────
    # STEP 2b: Fallback Search if RAG is weak or drug unknown
    # ─────────────────────────────────────────
    fallback_result = None
    if (not drug_info['matched']) or rag_is_weak(rag_results):
        print("🌐 Step 2b: Web Fallback Search...")
        fallback_result = web_fallback_search(working_query)
        results['fallback'] = fallback_result
        if verbose:
            print(f"  → Fallback used: {fallback_result.get('used')}")
            print(f"  → Reason: {fallback_result.get('reason')}")
    else:
        results['fallback'] = {"used": False, "reason": "rag_confident", "results": []}

    # ─────────────────────────────────────────
    # STEP 3: Safety Classification (Person 2)
    # ─────────────────────────────────────────
    print("🛡️  Step 3: Safety Classification...")
    safety_input = working_query

    if patient_profile:
        age = patient_profile.get('age', '')
        cond = patient_profile.get('chronic_conditions', [])
        kidney = patient_profile.get('kidney_function', '')
        liver = patient_profile.get('liver_function', '')

        safety_input = (
            f"عمر المريض: {age}. "
            f"أمراض: {', '.join(cond) or 'لا يوجد'}. "
            f"وظائف الكلى: {kidney}. "
            f"وظائف الكبد: {liver}. "
            f"السؤال: {working_query}"
        )

    safety_result = classify_safety(safety_input)
    results['safety'] = safety_result

    if verbose:
        print(f"  → Risk Level: {safety_result['risk_level']}")
        print(f"  → Probabilities: {safety_result['probabilities']}")
        print(f"  → {safety_result['explanation']}")

    # ─────────────────────────────────────────
    # STEP 4: Dosage (Person 3) — if dosage question and info complete
    # ─────────────────────────────────────────
    dosage_result = None

    if is_dosage_question and drug_info['matched'] and patient_profile:
        print("💊 Step 4: Dosage Calculation...")
        try:
            dosage_result = calculate_dosage(drug_info['drug_name'], patient_profile)
            results['dosage'] = dosage_result

            if verbose:
                print(f"  → Recommended Dose: {dosage_result.get('recommended_dose')}")
                print(f"  → Risk Category: {dosage_result.get('risk_category')}")
                print(f"  → Doctor Referral: {dosage_result.get('doctor_referral')}")

        except Exception as e:
            print(f"  → ⚠️  Dosage API error: {e}")
            results['dosage_error'] = str(e)

    else:
        if verbose:
            print("💊 Step 4: Dosage — skipped")

    # ─────────────────────────────────────────
    # STEP 5: Prompt Builder + Final LLM
    # ─────────────────────────────────────────
    print("🧠 Step 5: Building Final Response (Groq Llama)...")

    # Build context from RAG
    context_blocks = []
    for i, r in enumerate(rag_results, 1):
        context_blocks.append(
            f"[مصدر {i}]\n"
            f"الدواء: {r['drug_name']}\n"
            f"القسم: {r['section']}\n"
            f"النص: {r['text'][:600]}\n"
        )
    rag_context = "\n".join(context_blocks)

    # Build fallback context safely
    fallback_context = ""
    if fallback_result and fallback_result.get("used"):
        fb_blocks = []
        for i, item in enumerate(fallback_result.get("results", []), 1):
            title = item.get("title", "")
            url = item.get("url", "")
            content = item.get("content", "")[:500]

            fb_blocks.append(
                f"[Fallback {i}]\n"
                f"Title: {title}\n"
                f"URL: {url}\n"
                f"Content: {content}\n"
            )

        fallback_context = (
            "\n--- Web Fallback Context "
            "(للاستئناس فقط وليس بديلاً عن المصادر الرسمية) ---\n"
            + "\n".join(fb_blocks)
        )

    # Build dosage context
    dosage_context = ""
    if dosage_result:
        cot_str = "\n".join(dosage_result.get('chain_of_thought', []))
        dosage_context = f"""
--- تحليل الجرعة الشخصية من Person 3 ---
مستوى الخطورة حسب Person 3: {dosage_result.get('risk_category')}
الجرعة المقترحة: {dosage_result.get('recommended_dose', '')}
تحويل للطبيب: {dosage_result.get('doctor_referral')}
Safety Flag: {dosage_result.get('safety_flag')}
سبب التعديل: {dosage_result.get('dose_adjustment_reason', '')}
التحليل:
{cot_str}
""".strip()

    # Patient context
    patient_str = ""
    if patient_profile:
        patient_str = f"بيانات المريض: {json.dumps(patient_profile, ensure_ascii=False)}"

    high_risk = (
        str(safety_result.get('risk_level', '')).lower() == 'high'
        or bool(dosage_result and dosage_result.get('doctor_referral'))
    )

    final_prompt = f"""
سؤال المستخدم:
{working_query}

{patient_str}

الدواء المكتشف:
{drug_info.get('drug_name') if drug_info.get('matched') else 'غير محدد'}

نتائج RAG من قاعدة البيانات:
{rag_context}

{fallback_context}

نتيجة Safety Classifier:
- مستوى الخطر: {safety_result.get('risk_level')}
- التفسير: {safety_result.get('explanation')}
- الاحتمالات: {safety_result.get('probabilities')}

نتيجة Dosage Calculator:
{dosage_context if dosage_context else 'لا توجد نتيجة جرعة متاحة أو السؤال ليس عن جرعة.'}

مطلوب منك الآن:
- اكتب رد ودود وطبيعي بالعربية المصرية.
- لا تستخدم قالب ثابت.
- ابدأ بتلخيص بسيط للحالة لو بيانات المريض موجودة.
- لو الحالة High Risk أو يوجد doctor_referral=True، وضح بهدوء أن الدواء يحتاج مراجعة طبيب أو صيدلي قبل الاستخدام.
- لو الجرعة موجودة والحالة High Risk، قل إنها "نتيجة مبدئية من النظام" وليست تصريحًا بالاستخدام.
- اسأل 2 إلى 3 أسئلة متابعة مناسبة لحالة المريض.
- لا تخترع معلومة غير موجودة في السياق.
- لا تضف disclaimer في آخر الرد، سيتم إضافته تلقائياً.
""".strip()

    groq_client = Groq(api_key=GROQ_API_KEY)
    llm_resp = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": final_prompt}
        ],
        temperature=0.25,
        max_tokens=700,
    )

    final_response = llm_resp.choices[0].message.content.strip()

    # ─────────────────────────────────────────
    # STEP 6: Response Formatter
    # ─────────────────────────────────────────
    final_response = add_final_disclaimer(
            final_response,
            high_risk=high_risk,
            dosage_result=dosage_result
        )

    results['final_response'] = final_response

    print("=" * 70)
    print("✅ FINAL RESPONSE:")
    print("=" * 70)
    print(final_response)
    print("=" * 70)

    return results


print("✅ Main pipeline function ready!")
````

````python
# ============================================================
# RxChat Smart Chat Layer V2
# حط الخلية دي في آخر النوتبوك وشغل بعدها: run_rxchat_chat_v2()
# ============================================================

import json
import re
from io import StringIO
from contextlib import redirect_stdout
from groq import Groq

# ─────────────────────────────────────────
# 1) Override Final LLM personality
# ده بيأثر على rxchat_pipeline لأنها بتستخدم SYSTEM_PROMPT global
# ─────────────────────────────────────────

SYSTEM_PROMPT = """
أنت RxChat، مساعد صيدلي مصري ذكي وودود.

اتكلم مع المستخدم كأنه شات طبيعي، مش تقرير ثابت.
استخدم العربية المصرية البسيطة والواضحة.

قواعد أساسية:
- استخدم فقط معلومات RAG و Safety و Dosage الموجودة في السياق.
- لا تخترع جرعات أو معلومات طبية غير موجودة.
- لو السؤال عن جرعة والبيانات ناقصة، اسأل عن الناقص فقط.
- لو الحالة High Risk أو doctor_referral=True، لا تقول للمستخدم "خد الدواء".
- لو الجرعة موجودة في حالة خطرة، اذكرها كـ "نتيجة مبدئية من النظام" وليست تصريحًا بالاستخدام.
- اسأل 2 أو 3 أسئلة متابعة ذكية لو محتاج تفاصيل.
- لا تستخدم قالب ثابت كل مرة مثل: الاستخدام / الجرعة / التحذيرات / متى تذهب للطبيب.
- حتى لو طلب منك prompt الرد بشكل منظم، خليه منظم طبيعيًا بدون عناوين مكررة.
- لا تكرر disclaimer؛ النظام سيضيفه تلقائيًا.
- خليك مختصر ومفيد.
""".strip()


# ─────────────────────────────────────────
# 2) Chat Memory
# ─────────────────────────────────────────

chat_patient_profile = {}
chat_history = []
last_pending_query = None
debug_mode = False


def clean_chat_text(text: str) -> str:
    """Remove <think> blocks and clean text."""
    text = re.sub(r"<think>.*?</think>", "", text or "", flags=re.DOTALL | re.IGNORECASE)
    return text.strip()


def normalize_digits(text: str) -> str:
    """Convert Arabic digits to English digits."""
    arabic_digits = "٠١٢٣٤٥٦٧٨٩"
    english_digits = "0123456789"
    return text.translate(str.maketrans(arabic_digits, english_digits))


def update_patient_profile_from_text(text: str, profile: dict) -> dict:
    """
    Extract patient data from Egyptian Arabic text.
    مثال:
    عمره 8 سنين ووزنه 25 كيلو ومفيش أمراض والكلى طبيعية والكبد طبيعي
    """
    profile = profile or {}
    text = normalize_digits(text)
    lower = text.lower()

    # Age examples: عمره 8 / عمر 8 / 8 سنين / 70 سنة
    age_match = re.search(r"(?:عمره|عمرها|العمر|عمر|سن)\s*(\d+)", lower)
    if not age_match:
        age_match = re.search(r"(\d+)\s*(?:سنة|سنه|سنين)", lower)
    if age_match:
        profile["age"] = int(age_match.group(1))

    # Weight examples: وزنه 25 / وزن 25 / 25 كيلو
    weight_match = re.search(r"(?:وزنه|وزنها|الوزن|وزن)\s*(\d+)", lower)
    if not weight_match:
        weight_match = re.search(r"(\d+)\s*(?:كيلو|kg)", lower)
    if weight_match:
        profile["weight_kg"] = float(weight_match.group(1))

    # Gender
    if any(w in lower for w in ["ولد", "ذكر", "راجل", "رجل", "male"]):
        profile["gender"] = "ذكر"
    elif any(w in lower for w in ["بنت", "أنثى", "انثى", "ست", "female"]):
        profile["gender"] = "أنثى"

    # Chronic conditions
    conditions = set(profile.get("chronic_conditions", []))

    no_disease_phrases = [
        "مفيش أمراض", "مفيش امراض", "لا يوجد أمراض", "لا يوجد امراض",
        "مافيش أمراض", "مافيش امراض", "بدون أمراض", "بدون امراض",
        "معندوش أمراض", "معندوش امراض", "معندهاش أمراض", "معندهاش امراض"
    ]
    if any(p in lower for p in no_disease_phrases):
        conditions = set()

    if "ضغط" in lower:
        conditions.add("ضغط عالي")

    if "سكر" in lower or "سكري" in lower:
        conditions.add("سكري")

    if "قلب" in lower:
        conditions.add("مرض قلبي")

    if "قرحة" in lower or "نزيف معدة" in lower or "مشكلة معدة" in lower:
        conditions.add("قرحة أو مشكلة معدة")

    # Only add kidney/liver disease if not saying normal
    if ("كلى" in lower or "كلوي" in lower) and not any(p in lower for p in ["كلى طبيعية", "الكلى طبيعية", "وظائف الكلى طبيعية"]):
        if "قصور" in lower or "مشكلة" in lower or "مرض" in lower:
            conditions.add("مرض كلوي")

    if "كبد" in lower and not any(p in lower for p in ["كبد طبيعي", "الكبد طبيعي", "وظائف الكبد طبيعية"]):
        if "قصور" in lower or "مشكلة" in lower or "مرض" in lower:
            conditions.add("مرض كبدي")

    profile["chronic_conditions"] = list(conditions)

    # Kidney function
    if any(p in lower for p in ["الكلى طبيعية", "كلى طبيعية", "وظائف الكلى طبيعية", "kidney normal"]):
        profile["kidney_function"] = "normal"
    elif any(p in lower for p in ["قصور بسيط في الكلى", "قصور كلى بسيط", "مشكلة بسيطة في الكلى", "mild kidney"]):
        profile["kidney_function"] = "mild_impairment"
    elif any(p in lower for p in ["قصور متوسط في الكلى", "قصور كلى متوسط", "moderate kidney"]):
        profile["kidney_function"] = "moderate_impairment"
    elif any(p in lower for p in ["قصور شديد في الكلى", "قصور كلى شديد", "severe kidney"]):
        profile["kidney_function"] = "severe_impairment"

    # Liver function
    if any(p in lower for p in ["الكبد طبيعي", "كبد طبيعي", "وظائف الكبد طبيعية", "liver normal"]):
        profile["liver_function"] = "normal"
    elif any(p in lower for p in ["قصور بسيط في الكبد", "قصور كبد بسيط", "مشكلة بسيطة في الكبد", "mild liver"]):
        profile["liver_function"] = "mild_impairment"
    elif any(p in lower for p in ["قصور متوسط في الكبد", "قصور كبد متوسط", "moderate liver"]):
        profile["liver_function"] = "moderate_impairment"
    elif any(p in lower for p in ["قصور شديد في الكبد", "قصور كبد شديد", "severe liver"]):
        profile["liver_function"] = "severe_impairment"

    return profile


# ─────────────────────────────────────────
# 3) Smart Router
# ─────────────────────────────────────────

def quick_self_harm_check(text: str) -> bool:
    lower = text.lower()
    danger_words = [
        "انتحر", "انتحار", "اموت", "أموت", "اقتّل نفسي", "اقتل نفسي",
        "اذي نفسي", "أذي نفسي", "اخلص من حياتي", "جرعة قاتلة",
        "اخد علبة", "آخد علبة", "overdose", "اودّي نفسي"
    ]
    return any(w in lower for w in danger_words)


def ask_chat_router(user_msg: str, last_pending_query: str = None) -> dict:
    """
    LLM router:
    casual / medical / dosage / patient_info / self_harm / unclear
    """
    if quick_self_harm_check(user_msg):
        return {
            "intent": "self_harm",
            "confidence": 1.0,
            "reason": "self harm keywords detected"
        }

    client = Groq(api_key=GROQ_API_KEY)

    system = """
أنت Router ذكي داخل شات بوت صيدلي اسمه RxChat.
مهمتك تصنيف رسالة المستخدم فقط وليس الرد عليه.

اختار intent واحد فقط:
- casual: دردشة عادية مثل عامل ايه، ازيك، شكرا، تمام، صباح الخير، مساء الخير
- medical: سؤال عن دواء، أعراض جانبية، تداخلات، موانع استعمال، مرض أو عرض
- dosage: سؤال عن جرعة أو كام حبة أو مقدار دواء أو عدد مرات الاستخدام
- patient_info: المستخدم يرسل بيانات مريض مثل العمر، الوزن، وظائف الكلى، وظائف الكبد، أمراض مزمنة
- self_harm: المستخدم يتكلم عن الانتحار، إيذاء النفس، جرعة قاتلة، أو استخدام دواء لإيذاء نفسه
- unclear: الرسالة غير واضحة

قواعد:
- لو فيه سؤال عن جرعة → dosage
- لو فيه اسم دواء أو عرض أو مرض → medical
- لو المستخدم بيكمل عمر/وزن/أمراض/كلى/كبد فقط → patient_info
- لو كلام عادي أو هزار أو سلام → casual
- لو فيه إيذاء نفس → self_harm

ارجع JSON فقط بدون شرح:
{
  "intent": "casual|medical|dosage|patient_info|self_harm|unclear",
  "confidence": 0.0,
  "reason": "short reason"
}
"""

    user = f"""
رسالة المستخدم:
{user_msg}

هل يوجد سؤال سابق مستني بيانات؟
{bool(last_pending_query)}

السؤال السابق لو موجود:
{last_pending_query or ""}
"""

    resp = client.chat.completions.create(
        model="qwen/qwen3-32b",
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": user}
        ],
        temperature=0,
        max_tokens=150
    )

    txt = clean_chat_text(resp.choices[0].message.content)

    try:
        match = re.search(r"\{.*\}", txt, flags=re.DOTALL)
        if match:
            return json.loads(match.group(0))
        return json.loads(txt)
    except Exception:
        return {
            "intent": "medical",
            "confidence": 0.5,
            "reason": "router_json_parse_failed"
        }


# ─────────────────────────────────────────
# 4) Casual Chat Brain
# ─────────────────────────────────────────

def chat_brain_response(user_msg: str, history: list) -> str:
    """
    Natural chat brain for casual messages only.
    """
    client = Groq(api_key=GROQ_API_KEY)

    system = """
أنت RxChat، مساعد صيدلي مصري ودود وذكي.

اتكلم بالعربية المصرية الطبيعية، زي شات حقيقي.
رد في 2 إلى 4 سطور فقط.
خليك لطيف، واضح، ومفيد.

ممنوع:
- لا تذكر RAG أو Safety Classifier أو Pipeline للمستخدم.
- لا تعطي جرعات أو نصائح طبية تفصيلية هنا.
- لا تقول جمل غريبة أو مترجمة حرفيًا.
- لا تطول في الكلام.

أمثلة أسلوب:
User: عامل ايه؟
Assistant: تمام الحمد لله 😄 أنا RxChat، اسألني عن أي دواء أو جرعة أو أعراض جانبية وأنا أساعدك بأمان.

User: شكرا
Assistant: العفو يا فندم 😄 لو عندك أي سؤال عن دواء أو جرعة أنا معاك.

User: مش فاهم
Assistant: ولا يهمك، ابعتلي اسم الدواء أو السؤال اللي محيرك وأنا أوضحهولك ببساطة.
"""

    messages = [{"role": "system", "content": system}]

    for h in history[-4:]:
        messages.append(h)

    messages.append({"role": "user", "content": user_msg})

    resp = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        temperature=0.35,
        max_tokens=220
    )

    return clean_chat_text(resp.choices[0].message.content)


# ─────────────────────────────────────────
# 5) Crisis Response
# ─────────────────────────────────────────

def crisis_response() -> str:
    return (
        "أنا آسف إنك حاسس بكده. لو الكلام ده حقيقي ومش مجرد تجربة، مهم جدًا تبعد أي أدوية أو أدوات ممكن تأذيك دلوقتي، "
        "وتكلم شخص قريب منك فورًا أو تتواصل مع الطوارئ في بلدك.\n\n"
        "أنا مش هقدر أساعدك في أي جرعة ممكن تؤذيك، لكن أقدر أفضل معاك خطوة بخطوة. "
        "قولي بس: أنت في مكان آمن دلوقتي؟ وفي حد قريب منك تقدر تكلمه؟"
    )


# ─────────────────────────────────────────
# 6) Chat Loop
# ─────────────────────────────────────────

def run_rxchat_chat_v2():
    global chat_patient_profile, chat_history, last_pending_query, debug_mode

    print("💊 RxChat Chat Mode V2 started")
    print("اتكلم معاه عادي زي أي شات.")
    print("Commands: /exit | /profile | /reset | /debug")
    print("=" * 70)

    while True:
        user_msg = input("\nYou: ").strip()

        if not user_msg:
            continue

        if user_msg == "/exit":
            print("RxChat: تمام، مع السلامة 👋")
            break

        if user_msg == "/profile":
            print(json.dumps(chat_patient_profile, ensure_ascii=False, indent=2))
            continue

        if user_msg == "/reset":
            chat_patient_profile = {}
            chat_history = []
            last_pending_query = None
            print("RxChat: تمام، مسحت الذاكرة وبيانات المريض.")
            continue

        if user_msg == "/debug":
            debug_mode = not debug_mode
            print(f"Debug mode: {debug_mode}")
            continue

        # 1) Router decides message type
        route = ask_chat_router(user_msg, last_pending_query)
        intent = route.get("intent", "medical")

        if debug_mode:
            print(f"[Router] intent={intent} | reason={route.get('reason')}")

        # 2) Self-harm must not enter medical dosage pipeline
        if intent == "self_harm":
            reply = crisis_response()

            chat_history.append({"role": "user", "content": user_msg})
            chat_history.append({"role": "assistant", "content": reply})

            last_pending_query = None

            print("\nRxChat:")
            print(reply)
            continue

        # 3) Update patient profile from message
        chat_patient_profile = update_patient_profile_from_text(user_msg, chat_patient_profile)

        # 4) Casual message → casual brain only
        if intent == "casual" and not last_pending_query:
            reply = chat_brain_response(user_msg, chat_history)

            chat_history.append({"role": "user", "content": user_msg})
            chat_history.append({"role": "assistant", "content": reply})

            print("\nRxChat:")
            print(reply)
            continue

        # 5) If waiting for patient info, keep the old medical question
        if last_pending_query and intent in ["patient_info", "medical", "dosage", "unclear", "casual"]:
            query_to_send = last_pending_query
        else:
            query_to_send = user_msg

        # 6) Run medical pipeline
        try:
            if debug_mode:
                result = rxchat_pipeline(
                    query=query_to_send,
                    patient_profile=chat_patient_profile,
                    verbose=True
                )
            else:
                with redirect_stdout(StringIO()):
                    result = rxchat_pipeline(
                        query=query_to_send,
                        patient_profile=chat_patient_profile,
                        verbose=False
                    )

            reply = result.get("final_response", "مش قادر أطلع رد دلوقتي.")
            reply = clean_chat_text(reply)

            print("\nRxChat:")
            print(reply)

            chat_history.append({"role": "user", "content": user_msg})
            chat_history.append({"role": "assistant", "content": reply})

            # Store pending query if pipeline asks for missing info
            if result.get("status") == "need_more_info":
                last_pending_query = query_to_send
            else:
                last_pending_query = None

        except Exception as e:
            print("\nRxChat:")
            print(f"حصل خطأ أثناء تشغيل الشات: {e}")


print("✅ Smart Chat V2 ready. Run: run_rxchat_chat_v2()")
````

````python
# ============================================================
# PATCH: Strong quick router guard for casual messages
# حط الخلية دي بعد Smart Chat V2
# ============================================================

def quick_intent_guard(text: str) -> str | None:
    """
    Fast local guard before LLM router.
    بيرجع intent لو الرسالة واضحة جدًا.
    لو مش واضحة، يرجع None ويكمل للـ LLM router.
    """
    t = text.strip().lower()

    # normalize common Arabic variants
    t = t.replace("إ", "ا").replace("أ", "ا").replace("آ", "ا").replace("ى", "ي")

    casual_patterns = [
        "عامل اي", "عامله اي", "عامل اية", "عاملة اية",
        "ازيك", "ازي حضرتك", "اخبارك", "ايه الاخبار",
        "السلام عليكم", "صباح الخير", "مساء الخير",
        "هاي", "hi", "hello", "hey",
        "شكرا", "شكرًا", "تسلم", "تمام", "ماشي"
    ]

    if any(p in t for p in casual_patterns):
        return "casual"

    self_harm_patterns = [
        "انتحر", "انتحار", "اموت", "اقتلي نفسي", "اقتل نفسي",
        "اذي نفسي", "اخلص من حياتي", "جرعة قاتلة",
        "اخد علبة", "overdose"
    ]

    if any(p in t for p in self_harm_patterns):
        return "self_harm"

    dosage_patterns = [
        "جرعة", "كام حبة", "كم حبة", "مقدار", "dose",
        "كل كام ساعة", "اخد كام", "آخد كام"
    ]

    if any(p in t for p in dosage_patterns):
        return "dosage"

    patient_info_patterns = [
        "عمره", "عمرها", "وزنه", "وزنها", "كيلو",
        "الكلى", "كلى", "الكبد", "كبد",
        "ضغط", "سكر", "مفيش امراض", "مفيش أمراض",
        "لا يوجد امراض", "لا يوجد أمراض"
    ]

    if any(p in t for p in patient_info_patterns):
        return "patient_info"

    return None


def ask_chat_router(user_msg: str, last_pending_query: str = None) -> dict:
    """
    Patched router:
    1) quick local guard for obvious messages
    2) LLM router only if needed
    """

    quick_intent = quick_intent_guard(user_msg)
    if quick_intent:
        return {
            "intent": quick_intent,
            "confidence": 1.0,
            "reason": "quick_guard"
        }

    client = Groq(api_key=GROQ_API_KEY)

    system = """
أنت Router ذكي داخل شات بوت صيدلي اسمه RxChat.
مهمتك تصنيف رسالة المستخدم فقط وليس الرد عليه.

اختار intent واحد فقط:
- casual: دردشة عادية مثل عامل ايه، ازيك، شكرا، تمام، صباح الخير، مساء الخير
- medical: سؤال عن دواء، أعراض جانبية، تداخلات، موانع استعمال، مرض أو عرض
- dosage: سؤال عن جرعة أو كام حبة أو مقدار دواء أو عدد مرات الاستخدام
- patient_info: المستخدم يرسل بيانات مريض مثل العمر، الوزن، وظائف الكلى، وظائف الكبد، أمراض مزمنة
- self_harm: المستخدم يتكلم عن الانتحار، إيذاء النفس، جرعة قاتلة، أو استخدام دواء لإيذاء نفسه
- unclear: الرسالة غير واضحة

لو الرسالة دردشة عادية فقط، صنفها casual حتى لو مفيش دواء.
ارجع JSON فقط بدون شرح:
{
  "intent": "casual|medical|dosage|patient_info|self_harm|unclear",
  "confidence": 0.0,
  "reason": "short reason"
}
"""

    user = f"""
رسالة المستخدم:
{user_msg}

هل يوجد سؤال سابق مستني بيانات؟
{bool(last_pending_query)}

السؤال السابق لو موجود:
{last_pending_query or ""}
"""

    resp = client.chat.completions.create(
        model="qwen/qwen3-32b",
        messages=[
            {"role": "system", "content": system},
            {"role": "user", "content": user}
        ],
        temperature=0,
        max_tokens=150
    )

    txt = clean_chat_text(resp.choices[0].message.content)

    try:
        match = re.search(r"\{.*\}", txt, flags=re.DOTALL)
        if match:
            return json.loads(match.group(0))
        return json.loads(txt)
    except Exception:
        return {
            "intent": "unclear",
            "confidence": 0.5,
            "reason": "router_json_parse_failed"
        }


def run_rxchat_chat_v2():
    global chat_patient_profile, chat_history, last_pending_query, debug_mode

    print("💊 RxChat Chat Mode V2 started")
    print("اتكلم معاه عادي زي أي شات.")
    print("Commands: /exit | /profile | /reset | /debug")
    print("=" * 70)

    while True:
        user_msg = input("\nYou: ").strip()

        if not user_msg:
            continue

        if user_msg == "/exit":
            print("RxChat: تمام، مع السلامة 👋")
            break

        if user_msg == "/profile":
            print(json.dumps(chat_patient_profile, ensure_ascii=False, indent=2))
            continue

        if user_msg == "/reset":
            chat_patient_profile = {}
            chat_history = []
            last_pending_query = None
            print("RxChat: تمام، مسحت الذاكرة وبيانات المريض.")
            continue

        if user_msg == "/debug":
            debug_mode = not debug_mode
            print(f"Debug mode: {debug_mode}")
            continue

        # 1) Router
        route = ask_chat_router(user_msg, last_pending_query)
        intent = route.get("intent", "unclear")

        if debug_mode:
            print(f"[Router] intent={intent} | reason={route.get('reason')} | pending={last_pending_query}")

        # 2) Self-harm
        if intent == "self_harm":
            reply = crisis_response()
            chat_history.append({"role": "user", "content": user_msg})
            chat_history.append({"role": "assistant", "content": reply})
            last_pending_query = None

            print("\nRxChat:")
            print(reply)
            continue

        # 3) Casual ALWAYS goes to chat brain, even if there is old pending query
        # عشان لو قلت "عامل ايه" مايكملش سؤال طبي قديم بالغلط
        if intent == "casual":
            reply = chat_brain_response(user_msg, chat_history)

            chat_history.append({"role": "user", "content": user_msg})
            chat_history.append({"role": "assistant", "content": reply})

            print("\nRxChat:")
            print(reply)
            continue

        # 4) Update patient profile
        chat_patient_profile = update_patient_profile_from_text(user_msg, chat_patient_profile)

        # 5) If waiting for patient info, keep old medical question
        if last_pending_query and intent in ["patient_info", "medical", "dosage", "unclear"]:
            query_to_send = last_pending_query
        else:
            query_to_send = user_msg

        # 6) Medical pipeline
        try:
            if debug_mode:
                result = rxchat_pipeline(
                    query=query_to_send,
                    patient_profile=chat_patient_profile,
                    verbose=True
                )
            else:
                with redirect_stdout(StringIO()):
                    result = rxchat_pipeline(
                        query=query_to_send,
                        patient_profile=chat_patient_profile,
                        verbose=False
                    )

            reply = result.get("final_response", "مش قادر أطلع رد دلوقتي.")
            reply = clean_chat_text(reply)

            print("\nRxChat:")
            print(reply)

            chat_history.append({"role": "user", "content": user_msg})
            chat_history.append({"role": "assistant", "content": reply})

            if result.get("status") == "need_more_info":
                last_pending_query = query_to_send
            else:
                last_pending_query = None

        except Exception as e:
            print("\nRxChat:")
            print(f"حصل خطأ أثناء تشغيل الشات: {e}")


print("✅ Patch applied. Run /reset inside chat or restart, then run: run_rxchat_chat_v2()")
````

````python
# ============================================================
# PATCH: Show warning/disclaimer only when needed
# ============================================================

def response_contains_dose(text: str) -> bool:
    """
    Detect if the response actually contains dosage information.
    """
    text = text or ""
    dose_keywords = [
        "جرعة", "الجرعة", "مجم", "mg", "ملجم",
        "كل 4", "كل 6", "كل 8", "كل 12",
        "كل أربع", "كل ست", "كل ثمان",
        "حبة", "قرص", "ملعقة", "مرة يوميًا", "مرات يوميًا"
    ]
    return any(k in text for k in dose_keywords)


def add_final_disclaimer(final_response: str, high_risk: bool = False, dosage_result: dict = None) -> str:
    """
    Add safety warning/disclaimer only when needed:
    - High risk / doctor referral
    - Or actual dosage is returned
    """
    final_response = (final_response or "").strip()

    has_dosage_result = False
    if dosage_result:
        recommended_dose = dosage_result.get("recommended_dose", "")
        has_dosage_result = bool(recommended_dose)

    should_add_warning = high_risk
    should_add_disclaimer = high_risk or has_dosage_result or response_contains_dose(final_response)

    if should_add_warning:
        final_response += (
            "\n\n⚠️ **تنبيه مهم:** الحالة دي محتاجة مراجعة طبيب أو صيدلي قبل استخدام الدواء."
        )

    if should_add_disclaimer:
        final_response += (
            "\n\n---\n"
            "*المعلومات تعليمية فقط وليست بديلاً عن الطبيب أو الصيدلي.*"
        )

    return final_response


print("✅ Disclaimer patch ready")
````
