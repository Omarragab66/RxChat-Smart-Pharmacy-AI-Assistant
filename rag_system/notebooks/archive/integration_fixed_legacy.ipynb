## 🧪 Cell 3 — Person 2: Load Safety Classifier (AraBERT)

````python
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModelForSequenceClassification

print("⏳ Loading Safety Classifier (AraBERT)...")

id2label = {0: 'Low', 1: 'Medium', 2: 'High'}
label2id = {'Low': 0, 'Medium': 1, 'High': 2}

# tokenizer من الموديل الأصلي مش من checkpoint
safety_tokenizer = AutoTokenizer.from_pretrained(
    "aubmindlab/bert-base-arabertv02",
    use_fast=False
)

# model weights من checkpoint بتاعك
safety_model = AutoModelForSequenceClassification.from_pretrained(
    PATHS["safety_model"],
    num_labels=3,
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True
)

device = "cuda" if torch.cuda.is_available() else "cpu"
safety_model = safety_model.to(device)
safety_model.eval()

print(f"✅ Safety model loaded on: {device}")

def classify_safety(text: str) -> dict:
    """Person 2: تصنيف خطورة الاستعلام — Low / Medium / High"""
    encoding = safety_tokenizer(
        text, max_length=256, padding='max_length',
        truncation=True, return_tensors='pt'
    )
    encoding = {k: v.to(device) for k, v in encoding.items()}

    with torch.no_grad():
        outputs = safety_model(**encoding)
        probs   = F.softmax(outputs.logits, dim=-1)[0]

    pred_id    = probs.argmax().item()
    risk_level = id2label[pred_id]
    high_prob  = probs[2].item()

    flags = []
    if risk_level == 'High':
        flags       = ['DANGEROUS', 'CONSULT_DOCTOR']
        explanation = '⚠️ خطر طبي عالي — يرجى استشارة الطبيب فوراً'
    elif risk_level == 'Medium':
        flags       = ['CAUTION_REQUIRED']
        explanation = '⚡ احتياط مطلوب — راجع الصيدلاني'
    else:
        flags       = []
        explanation = '✅ آمن — لا توجد مخاطر واضحة'

    return {
        'risk_level':   risk_level,
        'safety_score': round(high_prob, 3),
        'risk_flags':   flags,
        'explanation':  explanation,
        'probabilities': {
            'low':    round(probs[0].item(), 3),
            'medium': round(probs[1].item(), 3),
            'high':   round(probs[2].item(), 3),
        }
    }


# Quick test
test = classify_safety("بانادول بيسبب نعاس؟")
print(f"\n🧪 Quick test: {test['risk_level']} | {test['probabilities']}")
````

## 💊 Cell 4 — Person 3: Load Dosage Calculator

````python
import json
import random
import joblib
from groq import Groq

print("⏳ Loading Dosage Calculator (DeepSeek + Rule-based)...")

# Load dosage data for few-shot
with open(PATHS["dosage_json"], "r", encoding="utf-8") as f:
    dosage_data = json.load(f)

# Load or build safety classifier
def build_features_simple(age, weight, gender, conditions, kidney, liver):
    kidney_map = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    liver_map  = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    return [
        age, weight,
        1 if gender == "ذكر" else 0,
        len(conditions),
        kidney_map.get(kidney, 0),
        liver_map.get(liver, 0),
        1 if any("سكري" in c for c in conditions) else 0,
        1 if any("قلب" in c or "ضغط" in c for c in conditions) else 0,
        1 if any("كبد" in c for c in conditions) else 0,
        1 if any("كلى" in c or "كلوي" in c for c in conditions) else 0,
    ]

def rule_based_risk(age, conditions, kidney, liver):
    """Rule-based risk — Person 3 DL component"""
    kidney_map = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    liver_map  = {"normal": 0, "mild_impairment": 1, "moderate_impairment": 2, "severe_impairment": 3}
    k = kidney_map.get(kidney, 0)
    l = liver_map.get(liver, 0)
    n = len(conditions)
    if k >= 3 or l >= 3 or n >= 3:
        return "HIGH"
    elif k >= 1 or l >= 1 or n >= 1 or age >= 65 or age <= 12:
        return "MEDIUM"
    return "LOW"


def calculate_dosage(drug_name: str, patient: dict) -> dict:
    """Person 3: حساب الجرعة الشخصية"""
    age        = patient.get("age", 30)
    weight     = patient.get("weight_kg", 70)
    gender     = patient.get("gender", "ذكر")
    conditions = patient.get("chronic_conditions", [])
    kidney     = patient.get("kidney_function", "normal")
    liver      = patient.get("liver_function", "normal")

    # Rule-based risk category
    risk_cat = rule_based_risk(age, conditions, kidney, liver)

    # Few-shot examples
    examples = random.sample(dosage_data, min(3, len(dosage_data)))
    few_shot = ""
    for ex in examples:
        pp  = ex.get("patient_profile", {})
        cot = "\n".join(ex.get("chain_of_thought", []))
        few_shot += f"""
=== مثال ===
الدواء: {ex.get('drug_name','')} | عمر: {pp.get('age','')} | وزن: {pp.get('weight_kg','')}
الأمراض: {', '.join(pp.get('chronic_conditions', [])) or 'لا يوجد'}
التحليل: {cot}
الجرعة: {ex.get('recommended_dose', 'يحتاج استشارة طبيب')}
"""

    cond_str   = ", ".join(conditions) if conditions else "لا يوجد"
    system_msg = f"""أنت نظام ذكاء اصطناعي متخصص في حساب جرعات الأدوية الشخصية.
قدّم تحليلاً خطوة بخطوة بناءً على بيانات المريض.
إذا كانت الحالة خطيرة قل: يجب تحويل للطبيب.
أمثلة:\n{few_shot}"""

    user_msg = f"""الدواء: {drug_name}
المريض: عمر {age} سنة | وزن {weight} كجم | جنس: {gender}
الأمراض المزمنة: {cond_str}
وظيفة الكلى: {kidney} | وظيفة الكبد: {liver}

قدّم التحليل خطوة بخطوة والجرعة الموصى بها:"""

    groq_client = Groq(api_key=GROQ_API_KEY)
    resp = groq_client.chat.completions.create(
        model="deepseek-r1-distill-llama-70b",
        messages=[
            {"role": "system", "content": system_msg},
            {"role": "user",   "content": user_msg}
        ],
        temperature=0.2,
        max_tokens=800,
    )

    raw      = resp.choices[0].message.content
    lines    = [l.strip() for l in raw.split("\n") if l.strip()]
    cot      = [l for l in lines if len(l) > 15][:5]
    ref_doc  = risk_cat == "HIGH" or "تحويل" in raw or "طبيب" in raw

    return {
        "drug_name":      drug_name,
        "risk_category":  risk_cat,
        "reasoning":      raw,
        "chain_of_thought": cot,
        "safety_flag":    ref_doc,
        "doctor_referral": ref_doc,
    }


print(f"✅ Dosage engine ready — {len(dosage_data)} examples loaded")
````

## 📚 Cell 5 — Person 4: Load RAG Pipeline

````python
import re
import numpy as np
import pandas as pd
from rapidfuzz import process, fuzz
from llama_index.core import Document, VectorStoreIndex, Settings
from llama_index.core.node_parser import SentenceSplitter
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

print("⏳ Loading RAG Pipeline...")

# Load CSV
df = pd.read_csv(PATHS["core_csv"])
df.columns = [str(c).strip() for c in df.columns]

needed_cols = [
    'drug_id', 'drug_name', 'generic_name', 'drug_class', 'form', 'route',
    'indications', 'dosage', 'mechanism', 'warnings', 'contraindications',
    'drug_interactions', 'adverse_reactions', 'top_adverse_events',
    'pregnancy_info', 'pediatric_info', 'geriatric_info', 'overdose_info',
    'storage_info', 'sources', 'is_egyptian', 'last_updated'
]
for col in needed_cols:
    if col not in df.columns:
        df[col] = ''

for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('').astype(str).str.strip()

print(f"✅ CSV loaded: {len(df)} drugs")


# ── Build structured section documents ──
def clean_text(x):
    x = str(x).strip()
    return re.sub(r'\s+', ' ', x) if x.lower() not in {'nan','none',''} else ''

def section_text(title_ar, title_en, content):
    c = clean_text(content)
    return f"{title_ar} / {title_en}:\n{c}" if c else ''

SECTIONS_MAP = [
    ('indications',        'الاستخدامات',           'Indications'),
    ('dosage',             'الجرعة',                'Dosage'),
    ('mechanism',          'آلية العمل',             'Mechanism'),
    ('warnings',           'التحذيرات',              'Warnings'),
    ('contraindications',  'موانع الاستخدام',        'Contraindications'),
    ('drug_interactions',  'التفاعلات الدوائية',     'Drug interactions'),
    ('adverse_reactions',  'الأعراض الجانبية',       'Adverse reactions'),
    ('top_adverse_events', 'أشهر البلاغات الجانبية', 'Top FAERS events'),
    ('pregnancy_info',     'الحمل والرضاعة',         'Pregnancy'),
    ('pediatric_info',     'الأطفال',                'Pediatric'),
    ('geriatric_info',     'كبار السن',              'Geriatric'),
    ('overdose_info',      'الجرعة الزائدة',         'Overdose'),
    ('storage_info',       'التخزين',                'Storage'),
]

documents = []
for _, row in df.iterrows():
    dname   = clean_text(row.get('drug_name', ''))
    generic = clean_text(row.get('generic_name', ''))

    # overview section
    header = f"""اسم الدواء: {dname} | Generic: {generic}
الفئة: {clean_text(row.get('drug_class',''))} | الشكل: {clean_text(row.get('form',''))}
Route: {clean_text(row.get('route',''))} | مصري: {'نعم' if row.get('is_egyptian')=='1' else 'غير مؤكد'}"""

    meta = {'drug_name': dname, 'generic_name': generic,
            'drug_class': clean_text(row.get('drug_class','')), 'section': 'overview'}
    documents.append(Document(
        text=f"دواء: {dname} | Generic: {generic} | Section: overview\n\n{header}",
        metadata=meta
    ))

    for col, ar, en in SECTIONS_MAP:
        text = section_text(ar, en, row.get(col, ''))
        if text:
            m = {**meta, 'section': col}
            documents.append(Document(
                text=f"دواء: {dname} | Generic: {generic} | Section: {col}\n\n{text}",
                metadata=m
            ))

print(f"✅ Built {len(documents)} section documents")


# ── Drug name matching ──
def normalize_ar(text):
    text = str(text).lower().strip()
    for old, new in [('[إأآا]','ا'),('ى','ي'),('ؤ','و'),('ئ','ي'),('ة','ه'),('[ًٌٍَُِّْـ]','')]:
        text = re.sub(old, new, text)
    return re.sub(r'\s+', ' ', re.sub(r'[^\w\s\u0600-\u06ff]', ' ', text)).strip()

name_rows = []
for _, row in df.iterrows():
    for field in ['drug_name', 'generic_name']:
        val = clean_text(row.get(field, ''))
        if val:
            name_rows.append({
                'candidate': val,
                'candidate_norm': normalize_ar(val),
                'drug_name': clean_text(row.get('drug_name', '')),
                'generic_name': clean_text(row.get('generic_name', '')),
            })

names_df = pd.DataFrame(name_rows).drop_duplicates(['candidate_norm', 'drug_name'])
choices  = names_df['candidate_norm'].tolist()


def detect_drug(query: str, threshold: int = 78) -> dict:
    q = normalize_ar(query)
    for _, r in names_df.iterrows():
        c = r['candidate_norm']
        if c and (c in q or q in c):
            return {'matched': True, 'drug_name': r['drug_name'],
                    'generic_name': r['generic_name'], 'score': 100, 'method': 'exact'}
    match = process.extractOne(q, choices, scorer=fuzz.WRatio)
    if match:
        norm, score, idx = match
        r = names_df.iloc[idx]
        if score >= threshold:
            return {'matched': True, 'drug_name': r['drug_name'],
                    'generic_name': r['generic_name'], 'score': round(float(score),2), 'method': 'fuzzy'}
    return {'matched': False, 'drug_name': None, 'generic_name': None, 'score': 0, 'method': 'none'}


# ── Build vector index ──
print("\n⏳ Building vector index (first run may take a few minutes)...")
Settings.embed_model = HuggingFaceEmbedding(model_name=EMBED_MODEL)
Settings.llm         = None
Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=80)

rag_index = VectorStoreIndex.from_documents(documents, show_progress=True)
print("✅ RAG index ready!")


# ── Retrieval function ──
def retrieve_context(query: str, top_k: int = 5) -> tuple:
    drug_info = detect_drug(query)
    if drug_info['matched']:
        subdocs = [d for d in documents if d.metadata.get('drug_name') == drug_info['drug_name']]
        if subdocs:
            temp_idx  = VectorStoreIndex.from_documents(subdocs)
            retriever = temp_idx.as_retriever(similarity_top_k=min(8, len(subdocs)))
        else:
            retriever = rag_index.as_retriever(similarity_top_k=top_k)
    else:
        retriever = rag_index.as_retriever(similarity_top_k=top_k)

    nodes = retriever.retrieve(query)
    results = [{
        'score':        float(n.score) if n.score else 0,
        'drug_name':    n.node.metadata.get('drug_name'),
        'generic_name': n.node.metadata.get('generic_name'),
        'section':      n.node.metadata.get('section'),
        'text':         n.node.text,
    } for n in nodes]

    return drug_info, results


# Quick test
di, res = retrieve_context("ما هي جرعة بانادول اطفال؟")
print(f"\n🧪 Drug detected: {di['drug_name']} | Sections: {[r['section'] for r in res]}")
````

## 🖼️ Cell 6 — Image Understanding (Groq Vision)

````python
import base64
from pathlib import Path
from groq import Groq


def image_to_base64(image_path: str) -> str:
    """تحويل الصورة لـ base64"""
    with open(image_path, "rb") as f:
        return base64.b64encode(f.read()).decode("utf-8")


def identify_drug_from_image(image_path: str) -> dict:
    """
    Person 4 — Vision Module
    يستقبل صورة دواء → يرجع اسم الدواء
    """
    ext = Path(image_path).suffix.lower()
    mime_map = {'.jpg': 'image/jpeg', '.jpeg': 'image/jpeg',
                '.png': 'image/png', '.webp': 'image/webp'}
    mime = mime_map.get(ext, 'image/jpeg')

    img_b64 = image_to_base64(image_path)

    groq_client = Groq(api_key=GROQ_API_KEY)
    resp = groq_client.chat.completions.create(
        model="meta-llama/llama-4-scout-17b-16e-instruct",  # Groq Vision model
        messages=[{
            "role": "user",
            "content": [
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:{mime};base64,{img_b64}"}
                },
                {
                    "type": "text",
                    "text": """أنت مساعد صيدلي.
انظر لصورة الدواء وأخبرني:
1. اسم الدواء التجاري
2. الاسم العلمي (generic name) إن ظهر
3. التركيز / الجرعة المكتوبة على العبوة إن ظهر

أجب بـ JSON فقط:
{"drug_name": "...", "generic_name": "...", "concentration": "...", "confidence": "high/medium/low"}"""
                }
            ]
        }],
        temperature=0.1,
        max_tokens=200,
    )

    raw = resp.choices[0].message.content.strip()

    # Parse JSON
    try:
        # Remove markdown code block if present
        raw_clean = re.sub(r'```json|```', '', raw).strip()
        result = json.loads(raw_clean)
    except:
        result = {'drug_name': raw, 'generic_name': '', 'concentration': '', 'confidence': 'low'}

    result['raw_response'] = raw
    return result


print("✅ Vision module ready")
print("\nUsage example:")
print('  result = identify_drug_from_image("path/to/drug_image.jpg")')
print('  print(result)')
````

## 🧠 Cell 7 — Main Pipeline: Everything Together

````python
SYSTEM_PROMPT = """
أنت RxChat، مساعد صيدلي تعليمي باللغة العربية المصرية.
قواعدك:
1) جاوب فقط من المعلومات الموجودة في السياق المرفق
2) لو المعلومة غير موجودة قل: لا أملك معلومة مؤكدة
3) لا تخترع جرعات أو تشخيصات
4) لو السؤال فيه خطر (جرعة زائدة، حمل، أطفال، فشل كلوي): انصح باستشارة طبيب
5) الرد منظم: الاستخدام، الجرعة، التحذيرات، متى تذهب للطبيب
6) في النهاية: المعلومات تعليمية وليست بديلاً عن الطبيب أو الصيدلي
""".strip()


def rxchat_pipeline(
    query: str,
    patient_profile: dict = None,
    image_path: str = None,
    verbose: bool = True
) -> dict:
    """
    🔗 Main RxChat Integration Pipeline
    =====================================
    Input:  query + optional patient_profile + optional image_path
    Output: final_response + all intermediate results
    """
    patient_profile = patient_profile or {}
    results = {}

    print("=" * 70)
    print(f"🔎 Query: {query}")
    if patient_profile:
        print(f"👤 Patient: {patient_profile}")
    print("=" * 70)

    # ─────────────────────────────────────────
    # STEP 0 (optional): Image → Drug Name
    # ─────────────────────────────────────────
    if image_path:
        print("\n📷 Step 0: Image Understanding...")
        vision_result = identify_drug_from_image(image_path)
        results['vision'] = vision_result

        # Inject drug name from image into query
        drug_from_img = vision_result.get('drug_name', '')
        if drug_from_img and drug_from_img.lower() not in query.lower():
            query = f"{query} [{drug_from_img}]"

        if verbose:
            print(f"  → Drug identified: {vision_result.get('drug_name')}")
            print(f"  → Generic: {vision_result.get('generic_name')}")
            print(f"  → Confidence: {vision_result.get('confidence')}")

    # ─────────────────────────────────────────
    # STEP 1: Drug Detection (Person 1 / Fuzzy)
    # ─────────────────────────────────────────
    print("\n🔍 Step 1: Drug Detection...")
    drug_info = detect_drug(query)
    results['drug_detection'] = drug_info

    if verbose:
        if drug_info['matched']:
            print(f"  → ✅ Detected: {drug_info['drug_name']} (score={drug_info['score']}, method={drug_info['method']})")
        else:
            print("  → ⚠️  No specific drug detected — global search")

    # ─────────────────────────────────────────
    # STEP 2: RAG Retrieval (Person 4)
    # ─────────────────────────────────────────
    print("\n📚 Step 2: RAG Retrieval...")
    _, rag_results = retrieve_context(query)
    results['rag'] = rag_results

    if verbose:
        for i, r in enumerate(rag_results, 1):
            print(f"  [{i}] {r['drug_name']} / {r['section']} (score={r['score']:.3f})")

    # ─────────────────────────────────────────
    # STEP 3: Safety Classification (Person 2)
    # ─────────────────────────────────────────
    print("\n🛡️  Step 3: Safety Classification...")
    safety_input = query
    if patient_profile:
        age  = patient_profile.get('age', '')
        cond = patient_profile.get('chronic_conditions', [])
        safety_input = f"عمر المريض: {age}. أمراض: {', '.join(cond) or 'لا يوجد'}. السؤال: {query}"

    safety_result = classify_safety(safety_input)
    results['safety'] = safety_result

    if verbose:
        print(f"  → Risk Level: {safety_result['risk_level']}")
        print(f"  → Probabilities: {safety_result['probabilities']}")
        print(f"  → {safety_result['explanation']}")

    # ─────────────────────────────────────────
    # STEP 4: Dosage (Person 3) — if dosage question
    # ─────────────────────────────────────────
    dosage_result = None
    dosage_keywords = ['جرعة', 'dose', 'كمية', 'كام حبة', 'مرات', 'مقدار']
    is_dosage_question = any(kw in query.lower() for kw in dosage_keywords)

    if is_dosage_question and drug_info['matched'] and patient_profile:
        print("\n💊 Step 4: Dosage Calculation...")
        try:
            dosage_result = calculate_dosage(drug_info['drug_name'], patient_profile)
            results['dosage'] = dosage_result
            if verbose:
                print(f"  → Risk Category: {dosage_result['risk_category']}")
                print(f"  → Doctor Referral: {dosage_result['doctor_referral']}")
        except Exception as e:
            print(f"  → ⚠️  Dosage API error: {e}")
    else:
        if verbose:
            print("\n💊 Step 4: Dosage — skipped (not a dosage question or missing patient data)")

    # ─────────────────────────────────────────
    # STEP 5: Prompt Builder + Final LLM
    # ─────────────────────────────────────────
    print("\n🧠 Step 5: Building Final Response (Groq Llama)...")

    # Build context from RAG
    context_blocks = []
    for i, r in enumerate(rag_results, 1):
        context_blocks.append(f"[مصدر {i}] الدواء: {r['drug_name']} | القسم: {r['section']}\n{r['text'][:600]}")
    rag_context = "\n\n".join(context_blocks)

    # Build dosage context
    dosage_context = ""
    if dosage_result:
        cot_str = "\n".join(dosage_result.get('chain_of_thought', []))
        dosage_context = f"""
--- تحليل الجرعة الشخصية ---
مستوى الخطورة: {dosage_result['risk_category']}
التحليل:
{cot_str}
"""

    # Patient context
    patient_str = ""
    if patient_profile:
        patient_str = f"\nبيانات المريض: {json.dumps(patient_profile, ensure_ascii=False)}"

    final_prompt = f"""
{rag_context}

{dosage_context}

معلومات السلامة: مستوى الخطر = {safety_result['risk_level']} | {safety_result['explanation']}
{patient_str}

سؤال المستخدم: {query}

اكتب رداً منظماً بالعربية المصرية:
""".strip()

    groq_client = Groq(api_key=GROQ_API_KEY)
    llm_resp = groq_client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",   "content": final_prompt}
        ],
        temperature=0.1,
        max_tokens=700,
    )

    final_response = llm_resp.choices[0].message.content

    # ─────────────────────────────────────────
    # STEP 6: Response Formatter
    # ─────────────────────────────────────────
    if safety_result['risk_level'] == 'High' or (dosage_result and dosage_result.get('doctor_referral')):
        final_response += "\n\n⚠️ **تحذير:** هذا الاستعلام يستدعي استشارة الطبيب أو الصيدلي فوراً."

    final_response += "\n\n---\n*المعلومات تعليمية فقط وليست بديلاً عن الطبيب أو الصيدلي*"

    results['final_response'] = final_response

    print("\n" + "=" * 70)
    print("✅ FINAL RESPONSE:")
    print("=" * 70)
    print(final_response)
    print("=" * 70)

    return results


print("✅ Main pipeline function ready!")
````
