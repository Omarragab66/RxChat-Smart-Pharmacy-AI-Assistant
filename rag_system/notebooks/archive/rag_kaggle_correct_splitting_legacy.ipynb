## 3) Find and load `Core_data.csv`

ارفع `Core_data.csv` كـ Kaggle Dataset أو ضيفه في نفس مجلد النوت بوك.
الخلية دي بتدور عليه تلقائيًا.

```python
def find_csv(filename='Core_data.csv'):
    candidates = []
    search_roots = [Path('/kaggle/input'), Path('/kaggle/working'), Path('.')]
    for root in search_roots:
        if root.exists():
            candidates.extend(root.rglob(filename))
    if not candidates:
        raise FileNotFoundError(
            f"Could not find {filename}. Upload it to Kaggle input dataset or /kaggle/working."
        )
    return candidates[0]

CSV_PATH = find_csv('Core_data.csv')
print('CSV_PATH =', CSV_PATH)

df = pd.read_csv(CSV_PATH)
print('Shape:', df.shape)
display(df.head(3))
```

## 4) Basic cleaning

مهم جدًا: لا تستخدم `df.head(10)` هنا. هنستخدم كل الأدوية.

```python
# Keep full dataframe, do NOT limit to head(10)
df = df.copy()

# Normalize column names
original_columns = df.columns.tolist()
df.columns = [str(c).strip() for c in df.columns]

# Make sure important text columns exist
needed_columns = [
    'drug_id', 'drug_name', 'generic_name', 'drug_class', 'form', 'route',
    'indications', 'dosage', 'mechanism', 'warnings', 'contraindications',
    'drug_interactions', 'adverse_reactions', 'top_adverse_events',
    'pregnancy_info', 'pediatric_info', 'geriatric_info', 'overdose_info',
    'storage_info', 'sources', 'is_egyptian', 'last_updated'
]

for col in needed_columns:
    if col not in df.columns:
        df[col] = ''

# Clean text values
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('').astype(str).str.strip()

# Numeric fallback
if 'data_quality_score' in df.columns:
    df['data_quality_score'] = pd.to_numeric(df['data_quality_score'], errors='coerce').fillna(0)

print('Rows/drugs:', len(df))
print('Unique drug names:', df['drug_name'].nunique())
print('Columns:', len(df.columns))
display(df[['drug_id', 'drug_name', 'generic_name', 'drug_class', 'form', 'data_quality_score']].head(10))
```

## 5) Build structured sections instead of one huge `rag_text`

بدل ما كل دواء يبقى نص واحد طويل، هنقسمه إلى sections صغيرة.
ده يحسن الـ retrieval جدًا لأن سؤال الجرعة يروح لقسم الجرعة، وسؤال التحذيرات يروح لقسم التحذيرات.

```python
def clean_text(x):
    if pd.isna(x):
        return ''
    x = str(x).strip()
    x = re.sub(r'\s+', ' ', x)
    return x


def section_text(title_ar, title_en, content):
    content = clean_text(content)
    if not content or content.lower() in {'nan', 'none', 'null'}:
        return ''
    return f"{title_ar} / {title_en}:{content}"


def build_drug_sections(row):
    drug_name = clean_text(row.get('drug_name', ''))
    generic = clean_text(row.get('generic_name', ''))
    drug_class = clean_text(row.get('drug_class', ''))
    form = clean_text(row.get('form', ''))
    route = clean_text(row.get('route', ''))
    sources = clean_text(row.get('sources', ''))
    last_updated = clean_text(row.get('last_updated', ''))
    is_egyptian = str(row.get('is_egyptian', '')).strip()

    header = f"""اسم الدواء: {drug_name}
الاسم العلمي: {generic}
الفئة الدوائية: {drug_class}
الشكل الدوائي: {form}
طريقة الاستخدام/Route: {route}
متوفر في مصر: {'نعم' if is_egyptian == '1' else 'غير مؤكد'}
المصادر: {sources}
آخر تحديث: {last_updated}
""".strip()

    sections = []
    sections.append(('overview', header))
    sections.append(('indications', section_text('الاستخدامات', 'Indications', row.get('indications', ''))))
    sections.append(('dosage', section_text('الجرعة وطريقة الاستخدام', 'Dosage', row.get('dosage', ''))))
    sections.append(('mechanism', section_text('آلية العمل', 'Mechanism', row.get('mechanism', ''))))
    sections.append(('warnings', section_text('التحذيرات', 'Warnings', row.get('warnings', ''))))
    sections.append(('contraindications', section_text('موانع الاستخدام', 'Contraindications', row.get('contraindications', ''))))
    sections.append(('interactions', section_text('التفاعلات الدوائية', 'Drug interactions', row.get('drug_interactions', ''))))
    sections.append(('side_effects', section_text('الأعراض الجانبية', 'Adverse reactions', row.get('adverse_reactions', ''))))
    sections.append(('faers_events', section_text('أشهر البلاغات الجانبية', 'Top FAERS adverse events', row.get('top_adverse_events', ''))))
    sections.append(('pregnancy', section_text('الحمل والرضاعة', 'Pregnancy', row.get('pregnancy_info', ''))))
    sections.append(('pediatric', section_text('الأطفال', 'Pediatric use', row.get('pediatric_info', ''))))
    sections.append(('geriatric', section_text('كبار السن', 'Geriatric use', row.get('geriatric_info', ''))))
    sections.append(('overdose', section_text('الجرعة الزائدة', 'Overdose', row.get('overdose_info', ''))))
    sections.append(('storage', section_text('التخزين', 'Storage', row.get('storage_info', ''))))

    # Drop empty sections
    sections = [(name, text) for name, text in sections if clean_text(text)]
    return sections


documents = []
for _, row in df.iterrows():
    for section_name, text in build_drug_sections(row):
        metadata = {
            'drug_id': clean_text(row.get('drug_id', '')),
            'drug_name': clean_text(row.get('drug_name', '')),
            'generic_name': clean_text(row.get('generic_name', '')),
            'drug_class': clean_text(row.get('drug_class', '')),
            'section': section_name,
        }
        # Put identifiers in text too, to help Arabic/English retrieval
        full_text = f"دواء: {metadata['drug_name']} | Generic: {metadata['generic_name']} | Section: {section_name}

{text}"
        documents.append(Document(text=full_text, metadata=metadata))

print('Number of structured documents:', len(documents))
print('Average document length:', int(np.mean([len(d.text) for d in documents])))
print('Max document length:', max(len(d.text) for d in documents))
print('
Example document:
')
print(documents[0].text[:1200])
```

## 6) Drug name matching

قبل الـ RAG، نحاول نعرف اسم الدواء من السؤال. لو عرفناه، نخلي البحث داخل هذا الدواء فقط.
ده أهم تعديل يخلي الديمو مستقر.

```python
def normalize_ar_text(text):
    text = str(text).lower().strip()
    text = re.sub(r'[إأآا]', 'ا', text)
    text = re.sub(r'ى', 'ي', text)
    text = re.sub(r'ؤ', 'و', text)
    text = re.sub(r'ئ', 'ي', text)
    text = re.sub(r'ة', 'ه', text)
    text = re.sub(r'[ًٌٍَُِّْـ]', '', text)
    text = re.sub(r'[^\w\s؀-ۿ]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Build candidates from Arabic trade name + English generic name
name_rows = []
for _, row in df.iterrows():
    for field in ['drug_name', 'generic_name']:
        value = clean_text(row.get(field, ''))
        if value:
            name_rows.append({
                'candidate': value,
                'candidate_norm': normalize_ar_text(value),
                'drug_name': clean_text(row.get('drug_name', '')),
                'generic_name': clean_text(row.get('generic_name', '')),
                'field': field,
            })

names_df = pd.DataFrame(name_rows).drop_duplicates(['candidate_norm', 'drug_name'])
choices = names_df['candidate_norm'].tolist()


def detect_drug(query, threshold=78):
    q_norm = normalize_ar_text(query)

    # Exact containment first
    for _, r in names_df.iterrows():
        cand = r['candidate_norm']
        if cand and (cand in q_norm or q_norm in cand):
            return {
                'matched': True,
                'drug_name': r['drug_name'],
                'generic_name': r['generic_name'],
                'matched_text': r['candidate'],
                'score': 100,
                'method': 'exact_contains'
            }

    # Fuzzy matching
    match = process.extractOne(q_norm, choices, scorer=fuzz.WRatio)
    if match:
        matched_norm, score, idx = match
        r = names_df.iloc[idx]
        if score >= threshold:
            return {
                'matched': True,
                'drug_name': r['drug_name'],
                'generic_name': r['generic_name'],
                'matched_text': r['candidate'],
                'score': round(float(score), 2),
                'method': 'fuzzy'
            }

    return {'matched': False, 'drug_name': None, 'generic_name': None, 'matched_text': None, 'score': 0, 'method': 'none'}

# Quick tests
for q in ['ما هي جرعة بانادول اطفال؟', 'ينفع اخد بروفين مع الضغط؟', 'اضرار كونجستال', 'panadol dosage']:
    print(q, '=>', detect_drug(q))
```

## 7) Embedding model and index

موديل embeddings متعدد اللغات مناسب للعربي.
أول تشغيل ممكن يأخذ وقت لأنه ينزل الموديل.

```python
# Good multilingual embedding for Arabic + English
# Small enough for Kaggle demo
EMBED_MODEL_NAME = 'sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2'

Settings.embed_model = HuggingFaceEmbedding(model_name=EMBED_MODEL_NAME)

# No global LLM here. We'll use Groq manually in a function if key exists.
Settings.llm = None

# Since documents are already section-based, we use small splitter mainly as a safety net.
Settings.node_parser = SentenceSplitter(chunk_size=512, chunk_overlap=80)

index = VectorStoreIndex.from_documents(documents, show_progress=True)
print('Index is ready ✅')
```

## 8) Retrieval function — drug-aware

لو السؤال فيه اسم دواء معروف: نبحث داخل documents الخاصة بالدواء فقط.
لو مفيش اسم واضح: نبحث global.

```python
def build_temp_index_for_drug(drug_name):
    subdocs = [d for d in documents if d.metadata.get('drug_name') == drug_name]
    if not subdocs:
        return None
    return VectorStoreIndex.from_documents(subdocs)


def retrieve_context(query, top_k=5, drug_top_k=8):
    drug_info = detect_drug(query)

    if drug_info['matched']:
        # Search only inside matched drug sections
        temp_index = build_temp_index_for_drug(drug_info['drug_name'])
        retriever = temp_index.as_retriever(similarity_top_k=drug_top_k)
        nodes = retriever.retrieve(query)
        scope = 'drug_specific'
    else:
        retriever = index.as_retriever(similarity_top_k=top_k)
        nodes = retriever.retrieve(query)
        scope = 'global'

    results = []
    for n in nodes:
        results.append({
            'score': float(n.score) if n.score is not None else None,
            'drug_name': n.node.metadata.get('drug_name'),
            'generic_name': n.node.metadata.get('generic_name'),
            'section': n.node.metadata.get('section'),
            'text': n.node.text,
        })

    return drug_info, scope, results


def show_retrieval(query):
    drug_info, scope, results = retrieve_context(query)
    print('Query:', query)
    print('Detected drug:', drug_info)
    print('Scope:', scope)
    print('=' * 80)
    for i, r in enumerate(results, 1):
        print(f"[{i}] score={r['score']:.4f} | drug={r['drug_name']} | section={r['section']}")
        print(r['text'][:700].replace('
', ' '))
        print('-' * 80)

show_retrieval('ما هي جرعة بانادول اطفال؟')
```

## 9) Prompt builder

هنا بنجبر الموديل يجاوب من الـ context فقط، ولو المعلومة ناقصة يقول مش متأكد.

```python
SYSTEM_RULES = """
أنت RxChat، مساعد صيدلي تعليمي باللغة العربية المصرية.
التزم بالقواعد التالية:
1) جاوب فقط من المعلومات الموجودة في السياق المرفق.
2) لو المعلومة غير موجودة، قل بوضوح: لا أملك معلومة مؤكدة كافية من البيانات المتاحة.
3) لا تخترع جرعات أو تشخيصات.
4) لو السؤال فيه خطر مثل جرعة زائدة، حمل، أطفال، مرض كبد/كلى، حساسية، أو تداخلات خطيرة: انصح باستشارة طبيب أو صيدلي فورًا.
5) اجعل الرد منظمًا ومختصرًا: الاستخدام، الجرعة إن وجدت، التحذيرات، متى تذهب للطبيب.
6) أضف في النهاية: المعلومات تعليمية وليست بديلًا عن الطبيب أو الصيدلي.
""".strip()


def build_prompt(query, retrieved_results, patient_profile=None):
    patient_profile = patient_profile or {}
    context_blocks = []
    for i, r in enumerate(retrieved_results, 1):
        context_blocks.append(
            f"[مصدر {i}] الدواء: {r['drug_name']} | القسم: {r['section']}
{r['text']}"
        )
    context = "

".join(context_blocks)

    prompt = f"""
{SYSTEM_RULES}

بيانات المريض إن وجدت:
{json.dumps(patient_profile, ensure_ascii=False, indent=2)}

السياق المسترجع من قاعدة بيانات الأدوية:
{context}

سؤال المستخدم:
{query}

اكتب الرد النهائي بالعربية المصرية:
""".strip()
    return prompt
```

## 10) Optional: Groq LLM answer

ضع مفتاح Groq في Kaggle Secrets باسم `GROQ_API_KEY`، أو اكتبه مؤقتًا في Environment Variable.

من Kaggle:
- Add-ons / Secrets
- Add secret: `GROQ_API_KEY`

لو المفتاح مش موجود، الخلية هتطبع الـ context فقط.

```python
# Try to read Groq key from Kaggle Secrets if available
try:
    from kaggle_secrets import UserSecretsClient
    user_secrets = UserSecretsClient()
    os.environ['GROQ_API_KEY'] = user_secrets.get_secret('GROQ_API_KEY')
except Exception:
    pass

GROQ_API_KEY = os.getenv('GROQ_API_KEY', '').strip()
print('Groq key found:', bool(GROQ_API_KEY))


def get_llm():
    if not GROQ_AVAILABLE:
        print('Groq integration package not available.')
        return None
    if not GROQ_API_KEY:
        print('No GROQ_API_KEY found. Retrieval will work, final LLM answer will be skipped.')
        return None
    return Groq(
        model='llama-3.3-70b-versatile',
        api_key=GROQ_API_KEY,
        temperature=0.1,
        max_tokens=700,
    )

llm = get_llm()
```

## 11) Main chat function

استخدم الدالة دي في الديمو.

```python
def rxchat_answer(query, patient_profile=None, print_context=False):
    drug_info, scope, results = retrieve_context(query)
    prompt = build_prompt(query, results, patient_profile=patient_profile)

    print('Detected drug:', drug_info)
    print('Retrieval scope:', scope)
    print('Retrieved sections:', [(r['drug_name'], r['section'], round(r['score'], 3)) for r in results])
    print('=' * 90)

    if print_context or llm is None:
        print('RETRIEVED CONTEXT PREVIEW')
        print('=' * 90)
        for i, r in enumerate(results, 1):
            print(f"[{i}] {r['drug_name']} / {r['section']} / score={r['score']:.4f}")
            print(r['text'][:1200])
            print('-' * 90)

    if llm is None:
        return 'لم يتم تشغيل LLM لأن GROQ_API_KEY غير موجود. الـ retrieval شغال فوق.'

    response = llm.complete(prompt)
    return str(response)
```

```python
questions = [
    'ما هي جرعة بانادول اطفال؟',
    'ينفع اخد بروفين مع الضغط؟',
    'ايه اعراض كونجستال الجانبية؟',
    'لو اخدت جرعة زيادة من بانادول اعمل ايه؟',
    'هل الدواء ده ينفع للحامل؟ بانادول',
]

for q in questions:
    print('

' + '#' * 100)
    print('QUESTION:', q)
    ans = rxchat_answer(q, patient_profile={'age': 30, 'weight': 75}, print_context=False)
    print('
ANSWER:
', ans)
```

## 14) Save retrieval test report

دي بتطلع ملف CSV بسيط يثبت في التقرير إن الـ RAG بيرجع sections صحيحة.

```python
test_rows = []
for q in questions:
    drug_info, scope, results = retrieve_context(q)
    for rank, r in enumerate(results, 1):
        test_rows.append({
            'query': q,
            'detected_drug': drug_info.get('drug_name'),
            'match_score': drug_info.get('score'),
            'scope': scope,
            'rank': rank,
            'retrieved_drug': r['drug_name'],
            'section': r['section'],
            'retrieval_score': r['score'],
            'text_preview': r['text'][:300]
        })

report_df = pd.DataFrame(test_rows)
report_path = '/kaggle/working/rag_retrieval_test_report.csv' if Path('/kaggle/working').exists() else 'rag_retrieval_test_report.csv'
report_df.to_csv(report_path, index=False, encoding='utf-8-sig')
print('Saved:', report_path)
display(report_df.head(20))
```

# ملاحظات مهمة للتسليم

لو الدكتور سألكم عملتوا إيه لتحسين الـ RAG:

1. لم نستخدم النص الطويل `rag_text` كقطعة واحدة.
2. قسمنا كل دواء إلى sections طبية واضحة: dosage, warnings, interactions, side effects...
3. أضفنا drug-name matching قبل الـ retrieval.
4. إذا تم التعرف على اسم الدواء، يتم البحث داخل هذا الدواء فقط لتقليل التشويش.
5. استخدمنا embedding متعدد اللغات مناسب للعربي والإنجليزي.
6. أضفنا prompt يمنع الهلوسة ويجبر الموديل يقول إنه لا يعرف عند نقص المعلومات.

الخلاصة: المشكلة لم تكن حجم الداتا، بل كانت في طريقة التقسيم واستخدام LLM ضعيف مثل TinyLlama.
